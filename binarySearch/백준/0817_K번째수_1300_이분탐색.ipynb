n이 최대 100,000 이기 때문에 탐색으로 풀면 시간초과 발생한다. 

|      |      |      |      |      | 5보다 작은 수 |
| ---- | ---- | ---- | ---- | ---- | ------------- |
| 1    | 2    | 3    | 4    | 5    | 5             |
| 2    | 4    | 6    | 8    | 10   | 2             |
| 3    | 6    | 9    | 12   | 15   | 1             |
| 4    | 8    | 12   | 16   | 20   | 1             |
| 5    | 10   | 15   | 20   | 25   | 1             |

k=5일 때, B[k] =  10 , i= 1~n
이때, 어떤 수보다 작거나 같은 수의 개수는 min(k/i, n)이다.
i * 1,  i * 2,  i * 3,  i * 4,  .....  , i * j 이기 때문에, k를 i로 나눈 몫이 그 행에서 작거나 같은 개수가 된다.
left =1과 right = n*n을 이용하여 mid (중간값)을 구하고 파라메트릭 서치(이분탐색)를 이용하여
k보다 작거나 같은 개수의 최적의 값을 구한다.

~~~python
if  count(mid) >= k:
        answer = mid
        right = mid -1
~~~

조건문에서 count(mid) ==k와 count(mid) >k를 분리하지 않는 이유는 
n = 3, k=5 인 경우, count(mid) ==k를 만족하지 못한다. 
그래서 이 문제는 최적화된 문제를 결정문제로 풀어서 해결하는  파라메트릭 서치 유형이라고 볼 수 있으며,
이분 탐색은 정렬된 배열에서 정확히 일치하는 값을 리턴한다면, 
파라메트릭 서치는 가장 좋은 답을 찾아서 리턴해 준다.

```python
n = int(input())
k = int(input())
def count(mid):
    cnt =0
    for i in range(1, n+1):
        cnt += min(mid//i, n)
    return cnt

left = 1 ;right = n*n
while left <= right:
    mid = (left +right)//2
    if  count(mid) >= k:
        answer = mid
        right = mid -1
    else:
        left = mid + 1
print(answer)
```
