```python
n, c = map(int, input().split())
arr = []
for i in range(n):
    arr.append(int(input()))
arr.sort()    
start =1
end= arr[-1] - arr[0]
result = 0

while start <= end:
    mid = (start + end)//2 # gap
    value = arr[0]
    cnt = 1
    # 현재의 mid 이용해 공유기 설치하기
    for i in range(1, n): # 앞에서 부터 설치해보기
        if arr[i] >= value + mid:
            value = arr[i]
            cnt += 1
    if cnt >= c: # c개 이상의 공유기 설치할 수 있는 경우, gap 증가
        start = mid +1
        result = mid # 최적의 결과 저장
    else: #c개의 공유기 설치할 수 없다면 gap 감소
        end = mid -1
        
print(result)
            
    
```
