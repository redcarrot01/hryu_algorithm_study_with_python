```python
n, c = map(int,input().split())
arr = []
for i in range(n):
    arr.append(int(input()))
arr.sort()

start  =1
end = arr[-1] - arr[0]
result = 0

while start <= end:
    mid = (start + end)//2 # gap
    temp =arr[0]
    cnt =1
    
    # 현재의 mid 값을 이용해 공유기를 설치
    for i in range(1, n):
        if arr[i] >= temp + mid:
            temp =arr[i]
            cnt += 1
    if cnt >= c:
        start = mid + 1
        answer = mid
    else:
        end = mid -1
print(answer)
```
