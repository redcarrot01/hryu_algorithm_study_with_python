```python
n, m = map(int, input().split())
arr = list(map(int, input().split()))

left, right = 0, 10001
while left <= right:
    mid = (left + right) // 2
    cnt = 1
    max_ = arr[0]
    min_ = arr[0]
    for i in range(1, n):
        if max_ < arr[i] :
            max_ = arr[i]
        if min_ > arr[i] :
            min_ = arr[i]
        if max_ - min_ > mid:
            min_ = arr[i]
            max_ = arr[i]
            cnt += 1
    if cnt > m:
        left = mid + 1
    else:
        right = mid -1
print(left)
    
```
