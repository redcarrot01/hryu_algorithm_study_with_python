```python
찾고자 하는 값이 중간점과 동일하다고 가정하여, 탐색을 수행한다.
중간점이 가리키는 값보다 중간점이 작다면 왼쪽 탐색, 중간점이 가리키는 값보다 중간점이 큰 경우 오른쪽 탐색
```

```python
def binary_search(array, start,end):
    if start > end:
        return None
    mid = (start + end) //2
    if array[mid] ==mid:
        return mid
    elif array[mid]> mid:
        return binary_search(array, start,mid-1)
    else:
        return binary_search(array,mid+1, end)
    
n = int(input())
array = list(map(int, input().split()))
index = binary_search(array,  0, n-1)

if index == None:
    print(-1)
else:
    print(index)
```
