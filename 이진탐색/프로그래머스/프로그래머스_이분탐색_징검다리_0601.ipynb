```python

내가 생각 해낸 부분:
바위간의 거리를 기준으로 한다
1 <=  최소거리 <= distance 
바위의 위치 정렬
start, end 위치 설정 -> 이부분이 명확하지 않았다

내가 생각 못한 부분:
깬 바위 수를 체크하여!! start, end 위치 설정해주기
하 .. 어려운건 아닌데 아침이라 그런가 머리가 안돌아갔다.. 흥

바위 깨는 조건 : 
mid : 바위 사이의 최소 거리
mid 보다 거리가 짧으면 깬다.
왜냐면 우리는 최솟값 중 최댓값을 구할 것이므로 

바위 사이의 최소거리를 정해놓고 그 최소거리보다 작은 바위들을 다 제거
만약 그 제거하는 바위의 갯수가 주어진 n보다 크다면 거리를 좁혀야 할 것이고, 
n보다 작다면 거리를 늘리기
이 때, 이 최소 거리를 찾는 방법은 당연히 이분탐색

https://velog.io/@hanturtle/%EC%9D%B4%EB%B6%84%ED%83%90%EC%83%89-%EC%A7%95%EA%B2%80%EB%8B%A4%EB%A6%AC
```

```python

def solution(distance, rocks, n):
    answer = 0
    rocks.sort()
    start = 1
    end = distance
    
    # 바위 모두 제거하는 경우
    if n == len(rocks):
        return distance

    while start <= end:
        mid = (start+end)//2
        cnt = 0
        j = -1
        
        for i in range(0, len(rocks)):
            length = 0
            
            # 현재 바위가 맨 앞 위치 
            if j == -1:
                length = rocks[i]
            else : # 그렇지 않다면 현재 위치 - 이전 위치
                length = rocks[i] - rocks[j]
            
            # 최소 거리 보다 짧다면 바위 깨기
            if length < mid:
                cnt +=1
            else:
                j = i
        if cnt > n:
            end = mid -1
        else:
            start = mid + 1
            answer = max(answer, mid)

    return answer
```
