2. 문자열 길이가 1 ,2 인 경우는 예외 처리
3. 2번에 해당되지 않는다면, 분할과 새로운 숫자 배열 생성 
- 분할 방법 : 문자열 길이가 3이상이면 3등분해야함
~~~
1| 2| 3| 4 |5 -> 3등분 하려면 작대기가 4곳 중 2곳에 위치해야한다.
따라서, 문자열 길이가 n이라면 (n-1)C2 의 경우의 수로 분할 가능하다.
~~~
- 분할 이후, 3등분한 문자열에서 홀수 개수를 구한 후 재귀

문자열 길이가 최대 9이기 때문에 시간복잡도 문제는 고려하지 않는다.

```python


n = input()
n = str(n)
min_cnt = int(1e9)
max_cnt = 0

def is_odd(n):
    cnt = 0
    for i in n:
        if int(i) % 2 !=0:
            cnt += 1
    return cnt


def solve(s, cnt):
    global min_cnt, max_cnt
    
    if len(s)==1:
        min_cnt = min(min_cnt, cnt)
        max_cnt = max(max_cnt, cnt)
    elif len(s) == 2:
        s = str( int(s[0]) + int(s[1])) 
        cnt += is_odd(s)
        solve(s,cnt)
    else:
        for i in range(0, len(s)-2):
            for j in range(i+1, len(s)-1):
                s1 = s[: i+1]
                s2 = s[i+1:j+1]
                s3 = s[j+1:]
                
                temp = str(int(s1) + int(s2)+ int(s3))
                solve(temp, is_odd(temp)+cnt)

solve(n, is_odd(n))
print(min_cnt, max_cnt)

```
