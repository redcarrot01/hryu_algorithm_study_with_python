```python
s = input()
n = len(s)
visited =['']*n
def sol(s, i):
    if not s:
        return
    temp = min(s)
    idx = s.index(temp)
    visited[idx+i] =temp
    print("".join(visited))
    sol(s[idx+1:],i+idx+1)
    sol(s[:idx], i)
sol(s, 0)
```

```python
s = input()
n = len(s)
visited = [0]*n

def sol(l, r):
    
    if l == r:
        return
    
    min_s = min(s[l:r])
    idx = s[l:r].index(min_s) + l
    
    visited[idx] = 1
    
    for i in range(n):
        if visited[i]==1:
            print(s[i], end="")
    print()
    sol(idx+1, r)
    sol(l, idx)
    
sol(0,n)
```
