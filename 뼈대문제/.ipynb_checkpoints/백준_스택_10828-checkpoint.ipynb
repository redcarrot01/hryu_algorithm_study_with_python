```python
n = int(input())
stack = []
for i in range(n):
    temp = input().split()
    command = temp[0]
    
    if command == "push":
        stack.append(temp[1])
    elif command == "pop":
        if stack :
            print(stack.pop())
        else:
            print(-1)
    elif command == "size":
        print(len(stack))
    elif command == "empty":
        if stack:
            print(0)
        else:
            print(1)
    elif command == "top":
        if stack:
            print(stack[-1])
        else:
            print(-1)
```

```python
n = int(input())
stack = []

def push(x):
    stack.append(x)

def pop():
    if stack:
        return stack.pop()
    else:
        return -1
    
def size():
    return len(stack)

def empty():
    return 0 if stack else 1

def top():
    return stack[-1] if stack else -1

for _ in range(n):
    temp = input().split()
    command = temp[0]
    
    if command == "push":
        push(temp[1])
    elif command == "pop":
        print(pop())
    elif command == "size":
        print(size())
    elif command == "empty":
        print(empty())
    elif command == "top":
        print(top())
```
