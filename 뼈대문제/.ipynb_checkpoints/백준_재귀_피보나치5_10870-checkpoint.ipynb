```python
# 재귀
n = int(input())
def fibo(n):
    if n == 0:
        return 0
    elif n ==1 or n ==2:
        return 1
    return fibo(n-1) + fibo(n-2)

print(fibo(n))
```

```python
# dp 메모제이션 + 재귀 : 탑다운
d = [0] * 21
n = int(input())
def fibo(n):
    if n == 0:
        return 0
    elif n ==1 or n ==2:
        return 1
    if d[n] != 0:
        return d[n]
    d[n] = fibo(n-1) + fibo(n-2)
    return d[n]

print(fibo(n))
```

```python
# dp 반복문 : 보텀업
d = [0]*21
d[0] = 0
d[1] = 1
d[2] = 1
n = int(input())
for i in range(3,n+1):
    d[i] = d[i-1] + d[i-2]
print(d[n])
```
