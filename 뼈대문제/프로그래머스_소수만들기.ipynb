```python
from itertools import permutaions

def solution(nums):
    cnt = 0
    result = list(itertools.permutations(nums, 3))
    for i in result :
        temp = 0
        for j in i:
            temp += j
        for k in range(2, temp):
            if temp % k == 0:
                break
            cnt += 1
        
    return cnt
```
