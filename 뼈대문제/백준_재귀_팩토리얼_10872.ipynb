```python
# 재귀
n = int(input())
def fac(n):
    if n==0:
        return 1
    elif n==1:
        return 1
    return n * fac(n-1)
print(fac(n))
```

```python
#라이브러리
import math
n = int(input())
if n==0:
    print(1)
else:
    print(math.factorial(n))
```

```python
#반복문 - 보텀업
n = int(input())
def fac(n):
    if n ==0:
        return 1
    answer =1
    for i in range(1, n+1):
        answer *= i
    return answer
print(fac(n))
```

```python
# 재귀 + 메모 : n에 해당하는 key 있으면 value 반환
dp = {}
n = int(input())
def fac(n):
    global dp
    if n in dp:
        return dp[n]
    elif n <= 1:
        return 1
    else:
        dp[n] = n * fac(n-1)
        return dp[n]
print(fac(n))
```
