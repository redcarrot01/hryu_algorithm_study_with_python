```python

n = int(input())
queue = []

for _ in range(n):
    temp = input().split()
    command = temp[0]
    if command == "push":
#         queue.insert(0, temp[1])
        queue.append(temp[1])
        print(queue)
    elif command == "pop":
        print(queue.pop(0) if queue else -1)
    elif command == "size":
        print(len(queue))
    elif command == "empty":
        print(0 if queue else 1)
    elif command == "front":
        print(queue[0] if queue else -1)
    elif command == "back":
        print(queue[-1] if queue else -1)

    
```
