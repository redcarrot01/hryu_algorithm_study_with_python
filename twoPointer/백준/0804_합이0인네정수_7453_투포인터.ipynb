```python
# 딕셔너리 풀이
n = int(input())
a, b, c, d = [0]*n, [0]*n, [0]*n, [0]*n
for i in range(n):
    a[i], b[i], c[i], d[i] = map(int, input().split())
arr1 = [0]*(n*n)
arr2 = {}
idx=0
cnt = 0
for i in range(n):
    for j in range(n):
        arr1[idx] = a[i] + b[j] 
        arr2[c[i] + d[j]] = arr2.get(c[i]+d[j], 0) + 1
        idx += 1
for i in arr1:
    i_ = -i
    cnt += arr2.get(i_, 0)
print(cnt)
```

```python
n = int(input())
a, b, c, d = [0]*n, [0]*n, [0]*n, [0]*n
for i in range(n):
    a[i], b[i], c[i], d[i] = map(int, input().split())
arr1 = []
arr2 = []
for i in range(n):
    for j in range(n):
        arr1.append(a[i] + b[j])
        arr2.append(c[i] + d[j])
arr1.sort()
arr2.sort(reverse=True)

ab_idx , cd_idx = 0, 0
cnt = 0
while ab_idx < n*n and cd_idx < n*n:
    temp = arr1[ab_idx] + arr2[cd_idx]
    if temp > 0:
        cd_idx += 1
    elif temp < 0:
        ab_idx += 1
    else :
        i = ab_idx
        j = cd_idx
        while  i < n*n and arr1[ab_idx] == arr1[i]:
            i += 1
        while j < n*n and arr2[cd_idx] == arr2[j]:
            j += 1
        cnt += (i- ab_idx)*(j- cd_idx)
        ab_idx = i
        cd_idx = j
print(cnt)
```
