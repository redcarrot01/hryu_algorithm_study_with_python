```python
n, m = map(int, input().split())
nlist = list(map(int, input().split()))

start =0; end = 0; cnt = 0

while end < len(nlist) and start < len(nlist):
    if sum(nlist[start:end+1]) < m:
        end += 1
    
    elif sum(nlist[start:end+1]) == m:
        cnt += 1
        end += 1
    else:
        start += 1
print(cnt)
```
