```python
# 맞는 코드
import sys
n = int(input())
arr = list(map(int, input().split()))
arr.sort()
min_ = int(1e9)
result = []
for i in range(n-2):
    temp = arr[i]
    left = i+1
    right = n-1
    while left < right :
        curr = temp + arr[left] + arr[right]
        if abs(curr) <= abs(min_):
            result = [temp,arr[left], arr[right]]
            min_ = temp + arr[left] + arr[right]
        if curr < 0:
            left += 1
        elif curr > 0:
            right -= 1
        else:
            print(result[0][1],result[0][2], result[0][3])
            sys.exit()  
print(result[0],result[1], result[2])
```
