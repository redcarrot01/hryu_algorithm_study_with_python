```python
n, d, k, c = map(int, input().split())
sushi = [0]*n
sushi = list(int(input()) for _  in range(n))
sushi += sushi[:k]
cnt =0
max_cnt =0
left, right = 0, k
while left < right and right < len(sushi):
    if sushi[left:right].count(c)==1:
        cnt = right-left
    elif sushi[left:right].count(c)==0 and sushi[right] == c:
        cnt = right-left+1
    
    left += 1
    right += 1
    max_cnt = max(cnt, max_cnt)
print(max_cnt)
```

```python
from collections import defaultdict

n, d, k, c = map(int, input().split())
sushi = [0]*n
sushi = list(int(input()) for _  in range(n))
sushi += sushi[:k-1]
set_ =defaultdict(int)
set_[c] += 1
start = 0
end = 0
max_cnt =0
for end in range(len(sushi)):
    set_[sushi[end]] += 1 
    # k개 보다 커진 경우
    if end >= k-1:
        max_cnt = max(max_cnt, len(set_))
        set_[sushi[start]] -= 1
        #print(set_)
        if set_[sushi[start]] == 0:
            del set_[sushi[start]]
        #print(set_)
        start += 1
print(max_cnt)
```
