```python
n, d, k, c = map(int, input().split())
sushi = [0]*n
sushi = list(int(input()) for _  in range(n))
sushi += sushi[:k]
cnt =0
max_cnt =0
left, right = 0, k
while left < right and right < len(sushi):
    if sushi[left:right].count(c)==1:
        cnt = right-left
    elif sushi[left:right].count(c)==0 and sushi[right] == c:
        cnt = right-left+1
    
    left += 1
    right += 1
    max_cnt = max(cnt, max_cnt)
print(max_cnt)
```
