```python
from collections import defaultdict

n, d, k, c = map(int, input().split())
sushi = [0]*n
sushi = list(int(input()) for _  in range(n))
sushi += sushi[:k-1]
set_ =defaultdict(int)
set_[c] += 1
start = 0
end = 0
max_cnt =0
for end in range(len(sushi)):
    set_[sushi[end]] += 1 
    # k개 보다 커진 경우
    if end >= k-1:
        max_cnt = max(max_cnt, len(set_))
        print(set_)
        set_[sushi[start]] -= 1
        if set_[sushi[start]] == 0:
            del set_[sushi[start]]
        print(set_)
        start += 1
print(max_cnt)
```
