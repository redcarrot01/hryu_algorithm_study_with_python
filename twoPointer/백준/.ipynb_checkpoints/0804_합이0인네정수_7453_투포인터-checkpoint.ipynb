```python
# 시간초과 뜨는 코드
# import sys
# input = sys.stdin.readline
n = int(input())
a, b, c, d = [0]*n, [0]*n, [0]*n, [0]*n
for i in range(n):
    a[i], b[i], c[i], d[i] = map(int, input().split())
arr1 = [0]*(n*n)
arr2 = [0]*(n*n)
idx = 0
for i in range(n):
    for j in range(n):
        arr1[idx] = a[i] + b[j] 
        arr2[idx] = c[i] + d[j]
        idx += 1
arr1.sort()
arr2.sort()
#print(arr1, arr2)
cnt =0
left, right = 0, n*n-1
while left <= n*n-1 and right >= 0:
    sum_arr = arr1[left] + arr2[right]
    if sum_arr < 0:
        left += 1
    elif sum_arr == 0:
        while arr1[left] + arr2[right] == 0:
            cnt += 1
            right -= 1
        left += 1
        right = n*n-1
    else:
        right -= 1
print(cnt)
```

```python
# 딕셔너리 풀이
n = int(input())
a, b, c, d = [0]*n, [0]*n, [0]*n, [0]*n
for i in range(n):
    a[i], b[i], c[i], d[i] = map(int, input().split())
arr1 = [0]*(n*n)
arr2 = {}
idx=0
cnt = 0
for i in range(n):
    for j in range(n):
        arr1[idx] = a[i] + b[j] 
        arr2[c[i] + d[j]] = arr2.get(c[i]+d[j], 0) + 1
        idx += 1
for i in arr1:
    i_ = -i
    cnt += arr2.get(i_, 0)
print(cnt)
```
