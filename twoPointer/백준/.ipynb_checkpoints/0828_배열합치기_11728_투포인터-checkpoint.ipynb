```python
n, m = map(int, input().split())

arr1 = list(map(int, input().split()))
arr2 = list(map(int, input().split()))

result = [0]*(n+m)
i=0; j=0; k=0

while i<n or j<m:
    # 리스트 arr2의 모든 원소가 처리되었거나 리스트 arr1의 원소가 더 작을 때
    if j>= m or (i<n and arr1[i]< arr2[j]):
        result[k] = arr1[i]
        i += 1
    # 리스트 arr1의 모든 원소가 처리되었거나 리스트 arr2의 원소가 더 작을 때
    else:
        result[k] = arr2[j]
        j += 1
    k += 1
for i in result:
    print(i, end= ' ')
```
