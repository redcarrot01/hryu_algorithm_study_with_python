```python
import sys
n, m = map(int, input().split())
arr = list(map(int, sys.stdin.readline().split()))
sum_value =0
prefix_sum = [0]

for i in arr:
    sum_value += i
    prefix_sum.append(sum_value)

for _ in range(m):
    k, l = map(int, sys.stdin.readline().split())
    print(prefix_sum[l]- prefix_sum[k-1])
```
