```python
# import sys
# input = sys.stdin.readline

s = input()
left_k =[]
right_k =[]
# 왼 -> 오른 방향으로 k 세기, R 나올 때마다 cnt를 저장
cnt =0
for i in s:
    if i == 'K':
        cnt += 1
    else:
        left_k.append(cnt)
print(left_k)
# 오 -> 왼 방향으로 k 세기, R 나올 때마다 cnt를 저장
cnt = 0
for i in s[::-1]:
    if i=='K':
        cnt += 1
    else:
        right_k.append(cnt)
print(right_k)
right_k.reverse()
print(right_k)
# 투포인터 
l, r = 0, len(left_k)-1
answer = 0
while l <= r:
    # r-l+1 , r의 총 개수 + 왼쪽 또는 오른쪽에 있는 k의 개수 중 작은 값
    answer = max(answer, r-l+1+2*min(left_k[l], right_k[r]))
    print(answer)
    # l의 왼쪽에 있는 k 개수와 r의 오른쪽에 있는 k의 개수보다 적으면 투포인터
    if left_k[l] < right_k[r]:
        l += 1
    else:
        r -= 1
print(answer)

# RRKKRRKKRRKRKRKRR
# [0, 0, 2, 2, 4, 4, 5, 6, 7, 7]
# [0, 0, 1, 2, 3, 3, 5, 5, 7, 7]
# [7, 7, 5, 5, 3, 3, 2, 1, 0, 0]
# 10
```
