```python
n = int(input())
arr = list(map(int, input().split())) # [-2, 4, -99, -1, 98]
arr.sort()
left, right = 0,n-1
result = []
while left <right:
    solution = arr[left] + arr[right]
    result.append((abs(solution), arr[left] ,arr[right]))
    if solution <=0:
        left += 1
        if left != right and abs(solution) > abs(arr[left] + arr[right]) :
            result.append((abs(arr[left] + arr[right]), arr[left] ,arr[right]))

    else:
        right -= 1
        if left != right and abs(solution) > abs(arr[left] + arr[right]) :
            result.append((abs(arr[left] + arr[right]), arr[left] ,arr[right]))

result.sort()
print(result[0][1], result[0][2])
```
