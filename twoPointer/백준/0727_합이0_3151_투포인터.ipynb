```python
# 메모리 초과
from itertools import combinations
n = int(input())
arr = list(map(int, input().split()))
print(arr)
cnt =0    
for combi in list(combinations(arr, 3)):
    if sum(combi) == 0:
        cnt+= 1
print(cnt)
```

```python
# 내코드
n = int(input())
arr = list(map(int, input().split()))
arr.sort()
cnt =0 

for i in range(n-2):
    left, right = i+1, n-1
    temp = -arr[i]
    j = n
    while left < right:
        sum_arr =arr[left]+arr[right]
        if sum_arr < temp:
            left += 1
        elif sum_arr == temp:
            if arr[left] == arr[right]:
                cnt += right - left
            else:  
                if j > right:
                    j = right
                    while j >=0 and arr[j-1] == arr[right]:
                        j-=1
                    cnt += right - j +1
            left += 1
        else:
            right -=1   
print(cnt)
# 125612 / 476
```

```python
# 다른 분 코드
n = int(input())
nums = list(map(int, input().split()))
ans = 0

dp = [0] * 40001
bias = 20000

for i in range(2, n):
    for j in range(i-2, -1, -1):
        dp[bias + nums[i-1] + nums[j]] += 1
    
    ans += dp[bias - nums[i]]
print(dp)
print(ans)
```
