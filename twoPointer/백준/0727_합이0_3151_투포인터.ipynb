```python
# 메모리 초과
from itertools import combinations
n = int(input())
arr = list(map(int, input().split()))
cnt =0    
for combi in list(combinations(arr, 3)):
    if sum(combi) == 0:
        cnt+= 1
print(cnt)
```

```python
n = int(input())
arr = list(map(int, input().split()))
arr.sort()
cnt =0 

for i in range(n-2):
    left, right = i+1, n-1
    temp = -arr[i]
    j = n
    while left < right:
        sum_arr =arr[left]+arr[right]
        if sum_arr < temp:
            left += 1
        elif sum_arr == temp:
            if arr[left] == arr[right]:
                cnt += right - left
            else:  
                if j > right:
                    j = right
                    while j >=0 and arr[j-1] == arr[right]:
                        j-=1
                    cnt += right - j +1
            left += 1
        else:
            right -=1   
print(cnt)
# 125612 / 476
```
