Queue()로 큐 만들기

```python
import queue

q = queue.Queue()
q.put("ss")
q.put(1)
q.qsize() # 2
q.get() # 'ss'
```

리스트로 큐 만들기

```python
q = list()
def enqueue(data):
    q.append(data)
def dequeue():
    data = q[0]
    del q[0]
    return data
for i in range(10):
    enqueue(i)
dequeue() # 0
```

```python
stack_one = []
stack_two = []
 
def _push(stack, data):
    stack.append(data)
    print(data)
 
def _pop(stack):
    data = stack[-1]
    del stack[-1]
    return data
 
def queue_by_two_stacks(q_data):
 
    ## (1) stack_one에 차례대로 push를 해준다.
    for i in range(q_data):  
        _push(stack_one, i)
    print('---')
    ## (2) stack_one에서 넣은 것을 나중에 넣은것부터 꺼내서(pop) stack_two에 넣는다
    for i in range(q_data):
        _push(stack_two, _pop(stack_one)) 
    print('---')
    ## (3) stack_two에 넣은 데이터를 나중에 넣은 것부터 꺼내서 출력한다.
    for i in range(q_data):
        print(_pop(stack_two))
queue_by_two_stacks(5)
# 0
# 1
# 2
# 3
# 4
# --- (1)
# 4
# 3
# 2
# 1
# 0
# --- (2)
# 0
# 1
# 2
# 3
# 4
```
