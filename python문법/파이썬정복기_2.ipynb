```python
song = "by the rivers of babylon.."
alphabet = dict()
for c in song:
    if c.isalpha() == False:
        continue
    c= c.lower()
    if c not in alphabet:
        alphabet[c] =1
    else:
        alphabet[c] += 1
print(alphabet)
# {'b': 3, 'y': 2, 't': 1, 'h': 1, 'e': 2, 'r': 2, 'i': 1, 'v': 1, 's': 1, 'o': 2, 'f': 1, 'a': 1, 'l': 1, 'n': 1}

# 방법 1. 키의 목록을 구해 정렬 
key= list(alphabet.keys())
key.sort()
for c in key:
    num = alphabet[c]
    print(c, "=>", num)
    
# 방법 2,알파벳 순회하여 조사
for code in range(ord('a'), ord('z')+1):
    c= chr(code)
    num = alphabet.get(c, 0)
    print(c, "=>", num)
# a => 1
# b => 3
# e => 2
# f => 1
# h => 1
# i => 1
# l => 1
# n => 1
# o => 2
# r => 2
# s => 1
# t => 1
# v => 1
# y => 2
```

```python
s1 = {1, 2, 3, 4, 5}
s2 = {4, 5, 6, 7}
print("합집합", s1|s2)
print("합집합",s2.union(s1))
print("교집합", s1&s2)
print("교집합",s2.intersection(s1))
print("차집합", s1-s2)
print("차집합",s2.difference(s1))
print("배타적차집합", s1^s2) # 한쪽 집합에만 있는 원소의 합
# 합집합 {1, 2, 3, 4, 5, 6, 7}
# 합집합 {1, 2, 3, 4, 5, 6, 7}
# 교집합 {4, 5}
# 교집합 {4, 5}
# 차집합 {1, 2, 3}
# 차집합 {6, 7}
# 배타적차집합 {1, 2, 3, 6, 7}
```
