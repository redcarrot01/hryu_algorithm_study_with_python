#### 출력

```python
# sep에 빈 문자열 지정하면, a와 b 두 변수 사이에 공백 없애고 붙여서 출력해준다.
a = 12
b = 34 
print(a, b, sep='') 
# 1234
```

```python
# end 인수는 자동 개행되는데, 다른 문자 지정도 가능하다. end인수에 ''(빈문자열) 지정하면 공백 출력한다.
a = 12
b = 34 
print(a, end='') 
print(b)
# 1234
```

```python
# sep은 내용사이에 구분자를, end는 문자열 마지막에 지정한다.
print(12, 34, 56, sep='^^', end='-->')
print(78)
# 12^^34^^56-->78
```

#### 입력

```python
# 괄호안에 질문 내용 적기 가능하고, 변수에는 입력받은 값이 저장된다.
price = input('가격 입력 : ')
num = input('개수 입력 : ')
sum = int(price) * int(num)
print('총액은', sum, '원입니다')
# 가격 입력 : 100
# 개수 입력 : 4
# 총액은 400 원입니다
```

```python
# 신형포맷팅
name = "ddbwls"
age = 25
height = 200
print("이름:{}, 나이:{}, 키:{}".format(name, age, height))
print("이름:{0}, 나이:{1}, 키:{2}".format(name, age, height))
print("이름:{name}, 나이:{age}, 키:{height}".format(name="하이", age=20, height=160))
# 이름:ddbwls, 나이:25, 키:200
# 이름:ddbwls, 나이:25, 키:200
# 이름:하이, 나이:20, 키:160
```

```python
n = int(input())
dp = [[0 for i in range(10)] for j in range(101)]
for i in range(1, 10):    
    dp[1][i] = 1
for i in range(2, n+1):    
    for j in range(10):        
        if j == 0:            
            dp[i][j] = dp[i-1][1]        
        elif j == 9:            
            dp[i][j] = dp[i-1][8]        
        else:            
            dp[i][j] = dp[i-1][j-1] + dp[i-1][j+1]
            print(sum(dp[n]) % 1000000000)
```
