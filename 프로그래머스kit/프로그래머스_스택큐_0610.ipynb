## 기능개발

```python
# 해결 1
def solution(progresses, speeds):
    answer = []
    length = len(progresses)
    count = []
    
    for i in range(length):
        if(100-progresses[i])> speeds[i]:
            if (100-progresses[i]) % speeds[i] ==0:
                count.append((100-progresses[i]) // speeds[i])
            else:
                count.append((100- progresses[i])//speeds[i]+1)
        else:
            count.append(1)
    #print(count)  [7, 3, 9]
    
    cnt =0
    
    p=count[0]
    for i in range(len(count)):
        if p< count[i]:
            answer.append(cnt)
            p = count[i]
            cnt = 0
        cnt += 1
    answer.append(cnt)
    
    return answer

solution([93, 30, 55],[1, 30, 5])
```

```python
# 해결 2 다른 분 코드
def solution(progresses, speeds):
    answer = []
    time = 0
    count = 0
    # 100 넘어가면 pop하고, 카운트해주기
    while len(progresses)>0:
        if(progresses[0] + time*speeds[0]) >= 100:
            progresses.pop(0)
            speeds.pop(0)
            count += 1
            
        else:
            # count 초기화해주기 
            if count > 0:
                answer.append(count)
                count =0
            # 100되기 전까지 time을 이용해서 시간 체크
            time += 1
    answer.append(count)
    return answer
solution([93, 30, 55],[1, 30, 5])
```

## 프린터

```python
'''
priorities 리스트는 요청 위치만 남을 때까지 pop한다!
요청 위치 고려 
1. 요청한 위치가 맨 앞에 있고, 요청 위치 뒤에 중요도 큰 문서 있는 경우
2. 요청 위치 맨 앞이 아니고, 요청 위치 뒤에 중요도 큰 문서 있는 경우
'''
```

```python
def solution(priorities, location):
    count = 0
    while(len(priorities)!= 0):
        if location == 0:
            if priorities[0] < max(priorities): 
                priorities.append(priorities.pop(0))
                location = len(priorities)-1
            else:
                return count +1
        else:
            if priorities[0] < max(priorities):
                priorities.append(priorities.pop(0))
                location -= 1
            else:
                priorities.pop(0)
                location -= 1
                count += 1
    return count
solution([2,1,3,2,1],1)
```

## 다리를 지나는 트럭

```python
'''
bridge : 다리 통과 중인 트럭  
bridge_length : 다리 길이  
weight : 무게  
truck_weights : 트럭 별 무게  
bridge는 다리 길이 만큼 크기를 갖고, 이 크기를 유지할 필요가 있다.
bridge가 비어있지 않다면, bridge의 맨앞 원소 pop, 대기 트럭 살펴서 조건에 맞게 수행한다.
bridge가 비어있다면, 이미 대기트럭은 비어있다.
'''
```

```python
def solution(bridge_length, weight, truck_weights):
    cnt =0
    bridge = [0]*bridge_length
    
    while bridge:
        # 맨 앞 pop
        bridge.pop(0)
        cnt += 1
        # 대기하고 있는 트럭
        if truck_weights:
            sum_bridge = sum(bridge)
            if(sum_bridge + truck_weights[0]) <= weight:
                bridge.append(truck_weights.pop(0))
            else:
                bridge.append(0)
        
    return cnt

solution(2, 10, [7,4,5,6])
```

## 주식가격

```python
def solution(prices):
    answer = []
    for i in range(len(prices)):
        cnt =0
        for j in range(i+1, len(prices)):
            cnt += 1
            if prices[i] > prices[j]:
                break
        answer.append(cnt)
    
    return answer
solution([1, 2, 3, 2, 3])
```
