```python
#import sys
#input = sys.stdin.readline
import heapq
heap = []
n = int(input())
for i in range(n):
    x = int(input())
    if x == 0:
        if not heap:
            print(0)
        else:
            y = heapq.heappop(heap)[1]
            print(y)
    else:
        heapq.heappush(heap, (-x, x))
        
```
