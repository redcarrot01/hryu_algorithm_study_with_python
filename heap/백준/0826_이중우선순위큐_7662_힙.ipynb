```python
# 시간 초과 뜬느 풀이
import heapq
t = int(input())
for k in range(t):
    n = int(input())
    heap = []
    for i in range(n):
        command, num = input().split()
        num = int(num)
        if command == 'I':
            heapq.heappush(heap, num)
        else: 
            if len(heap) ==0:
                pass
            elif num == 1:
                heap = heapq.nlargest(len(heap), heap)[1:]
                heapq.heapify(heap)
            elif num == -1:
                heapq.heappop(heap)
        #print(heap)    
    if heap:
        max_heap = []
        max_heap = heapq.nlargest(1, heap)
        print(heapq.heappop(max_heap), heapq.heappop(heap))
    else:
        print('EMPTY')
```
