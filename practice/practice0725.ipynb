```python
1. 문자열 길이가 1일때 홀수의 개수 min, max 구하기
2. 문자열 길이가 2일때 서로 더해서 홀수의 개수를 세고, 문자열길이가 1될때까지 반복
3. 문자열 길이가 3이상일때,
- 3등분 분할
- 분할한 수 더해서, 홀수의 개수 구하기
- 문자열길이가 1될때까지 반복
```

```python
n = input()
min_cnt = int(1e9)
max_cnt = 0

def is_odd(n):
    cnt = 0
    for i in n:
        if int(i) % 2 != 0:
            cnt += 1
    return cnt
            

def solve(s, cnt):
    global min_cnt, max_cnt
    
    if len(s) == 1:
        min_cnt = min(min_cnt, cnt)
        max_cnt = max(max_cnt, cnt)
    elif len(s) == 2:
        s = str( int(s[0]) + int(s[1]) )
        cnt += is_odd(s)
        solve(s, cnt)
    else:
        for i in range(0, len(s)-2):
            for j in range(i+1, len(s)-1):
                s1 = s[:i+1]
                s2 = s[i+1:j+1]
                s3 = s[j+1:]
                temp = str(int(s1)+int(s2)+int(s3))
                solve(temp, is_odd(temp)+cnt)
    
solve(n, is_odd(n))
print(min_cnt, max_cnt)
```
