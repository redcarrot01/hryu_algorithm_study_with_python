```python
# 2
from itertools import permutations
def solution(k, m):
    cnt =0
    arr = [str(i) for i in range(1, k+1)]
    #print(arr)
    for combi in list(permutations(arr, k)):
        combi = list(combi)
        if int(''.join(combi)) % m == 0:
            cnt += 1   
    return cnt
solution(9,1000000002)
```

```python
# 1
from collections import defaultdict
def solution(color, prices):
    answer =0
    piece_cnt =0
    case = ['R', 'G', 'W', 'B', 'Y']
    dic1 = defaultdict(int)
    dic2 = defaultdict(int)
    for i in range(len(color)):
        dic1[color[i][0]] += 1
        dic2[color[i][1]] += 1
    for i in case:
        piece_cnt += abs(dic1[i] - dic2[i])
        print(piece_cnt)
    piece_cnt //= 2
    answer = (len(color)-piece_cnt)*prices[0] + piece_cnt*prices[1]
    #print(piece_cnt)
    answer1, answer2 = 0,0
    if prices[0] < prices[1]//2 and piece_cnt*2 <= prices[1]//prices[0]:
        answer1 =  (len(color)-piece_cnt)*prices[0] + piece_cnt*2*prices[0]
    #answer2 = len(color)*prices[1]
    answer = min(answer,  len(color)*prices[1])
    if answer1 :
        answer = min(answer, answer1)
    return answer
#solution(["RG", "WR", "BW", "GG"], [5000, 6000])
solution(["RG", "WR", "BW", "GG"], [2000, 6000])
#solution(['BW','RY', 'BY'], [9000, 10000])
```

```python
# 3
def solution(l1, l2):
    answer = []
    if l1 > l2:
        l1, l2 = l2, l1
    answer.append(l1)
    answer.append(l2)
    l2_temp =0
    while l2_temp < l2:
        l2_temp += l1
        if l2_temp > l2:
            break
        answer.append(l2_temp)
    answer = list(set(answer))
    answer.sort()
    return answer
solution(3, 8)
solution(2, 4)
solution(6, 1000)
```
