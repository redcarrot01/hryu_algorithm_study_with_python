```python
# 배열 돌리기1
'''
배열을 r만큼 반시계 방향으로 돌리기  
1. 테두리 개수 세기 -> cnt = min(n,m) //2  
2. 반복문을 통해 테두리 개수 만큼 회전한다.  
- 회전 시작하는 지점은 다른 변수에 저장  
- 다른 곳은 반시계 방향 회전
- 회전 시작 지점 회전 반영  
'''
n, m, r = map(int, input().split())
graph = [list(map(int, input().split())) for _ in range(n)]

def solution():
    for _ in range(r):
        rotate()
    for i in graph:
        for j in i:
            print(j, end=' ')
        print()
        
def rotate():
    cnt = min(n,m) // 2
    for i in range(cnt):
        nn = n - i -1
        mm = m - i -1
        start= graph[i][i]
        # 윗변
        for j in range(i,mm):
            graph[i][j] = graph[i][j+1]
        # 오른쪽 변
        for j in range(i, nn):
            graph[j][mm] = graph[j+1][mm]
        # 밑변
        for j in range(mm, i, -1):
            graph[nn][j] = graph[nn][j-1]
        # 왼쪽 변
        for j in range(nn, i, -1):
            graph[j][i] = graph[j-1][i]
        graph[i+1][i] = start
solution()
```

```python
# 배열 돌리기2
'''
배열 돌리기 1 같은 경우, 회전 횟수 만큼 탐색한다. 배열 돌리기 2 에서는 시간 복잡도를 낮춰야 한다.
예를 들어, 5x5 배열이 11바퀴를 돌린 것은 1바퀴를 돌린 것과 같으므로, 여기서는 불필요한 회전을 하지 않고, 
deque를 사용하여 해결한다.
deque를 사용하여 시간 복잡도를 낮춘다. 
'''
from collections import deque

def rotate(x, y, height, width):
    q = deque()
    # 윗변
    for i in range(y, y+width):
        q.append(graph[x][i])
    # 오른쪽
    for i in range(x+1, x+height):
        q.append(graph[i][y+width-1])
    # 밑변
    for i in range(y+width-2, y, -1):
        q.append(graph[x+height-1][i])
    # 왼쪽
    for i in range(x+width-1, x, -1):
        q.append(graph[i][y])
        
    q.rotate(-r)
    
    for i in range(y, y+width):
        graph[x][i] = q.popleft()
        
    for i in range(x+1, x+height):
        graph[i][y+width-1] = q.popleft()
    
    for i in range(y+width-2, y, -1):
        graph[x+height-1][i] = q.popleft()
    
    for i in range(x+height-1, x, -1):
        graph[i][y] = q.popleft()    
    
n, m, r = map(int, input().split())
graph = [list(map(int, input().split())) for _ in range(n) ]

height = n
width = m
x, y = 0, 0

while True:
    if height == 0 or width == 0:
        break
    rotate(x, y, height, width)
    x += 1 ; y += 1 ; height -= 2 ; width -= 2

    
for i in graph:
    for j in i:
        print(j, end= " ")
    print()
    
```
