```python
n,m,k= map(int, input().split())
arr = list(map(int, input().split()))
arr.append(0)
arr.append(k)
arr.sort()
left = 1
right = k-1

def count(length):
    cnt =0
    for i in range(1, n+2):
        cnt += (arr[i]- arr[i-1]-1)//length
    return cnt
while left <= right:
    mid = (left+right)//2
    if count(mid) <= m:
        result = mid
        right = mid-1
    else:
        left = mid + 1
print(result)
```
