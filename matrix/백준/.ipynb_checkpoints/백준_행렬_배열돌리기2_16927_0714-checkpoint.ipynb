배열돌리기1과 같은 문제 , deque를 사용하여 시간복잡도를 낮춘다. 
배열돌리기1 코드 같은 경우 5x5 배열에서 11바퀴를 돌린것은 1바퀴를 돌린 것과 같은데, 불필요한 회전을 반복한다.
따라서, 여기서는 불필요한 회전을 줄이고, deque를 사용하여 해결해보았다.  
~~~
배열돌리기1
2 ≤ N, M ≤ 300  
1 ≤ R ≤ 1,000
min(N, M) mod 2 = 0
1 ≤ Aij ≤ 108

배열돌리기2
2 ≤ N, M ≤ 300
1 ≤ R ≤ 10^9
min(N, M) mod 2 = 0
1 ≤ Aij ≤ 108
~~~

###  deque
양방향 큐  
큐보다 append와 pop이 압도적으로 빠르다. 
양끝 요소에 접근하여 삽입 삭제 하는 경우, 리스트는 o(n) , 데큐는 o(1) 로 가능  
~~~
from collections import deque

deq = deque()

# Add element to the start
deq.appendleft(10)

# Add element to the end
deq.append(0)

# Pop element from the start
deq.popleft()

# Pop element from the end
deq.pop()
~~~
- rotate 메서드 : 양수 또는 음수값ㅇㄹ 파라미터로 주어 좌, 우 회전 가능  
1 : 좌 -> 우로 -1: 우 -> 좌 
~~~
# Contain 1, 2, 3, 4, 5 in deq
deq = deque([1, 2, 3, 4, 5])
deq.rotate(1)
print(deq)
# deque([5, 1, 2, 3, 4])
deq.rotate(-1)
print(deq)
# deque([1, 2, 3, 4, 5])
~~~
- 필요성
리스트보다 월등한 연산 속도 제공  

```python
# 다른 이 풀이
from collections import deque

n, m, r = map(int, input().split())  
graph = [list(map(int, input().split())) for _ in range(n)]

def rotate(y, x, height, width):
    global graph
    q = deque()
    # 윗 변 : 왼 -> 오른쪽
    for i in range(x, x+width):
        q.append(graph[y][i])
    # 오른쪽 변 : 위 -> 아래
    for i in range(y+1, y+height):
        q.append(graph[i][x+width-1])
    # 밑 변 : 오른 -> 왼
    for i in range(x+width-2, x, -1):
        q.append(graph[y+height-1][i])
    # 왼쪽 변 : 아래 -> 위
    for i in range(y+height-1, y, -1):
        q.append(graph[i][x])
    # r만큼 큐가 오른 -> 왼쪽으로 당겨진다.
    q.rotate(-r) 
    
    for i in range(x, x+ width):
        graph[y][i] = q.popleft()
    
    for i in range(y+1, y+height):
        graph[i][x+width-1] = q.popleft()
    
    for i in range(x+width-2, x, -1):
        graph[y+height-1][i] = q.popleft()
    
    for i in range(y+height-1, y, -1):
        graph[i][x] = q.popleft()
height = n
width = m
x, y = 0, 0

while True:
    if height == 0 or width == 0: break
    
    rotate (y, x, height, width) # 세로 가로 세로 가로
    y += 1
    x += 1
    height -= 2
    width -= 2
    
for i in range(n):
    for j in range(m):
        print(graph[i][j], end = " ")
    print()
```

```python
# 참고한 내 풀이
# 출력문을 뒤에 배치해야 런타임 에러 안남 
# 이유는 ? NameError는 침조하지 않은 변수 나타나면 발생 : rotate를 뒤에서 앞으로 호출하는 불편함? 
# so 고침 코드 
from collections import deque

def rotate(x, y, height, width):
    #global graph
    q = deque()
    # 윗변 왼 -> 오른
    for i in range(y, y+width):
        q.append(graph[x][i])
    # 오른변 위 -> 아래
    for i in range(x+1, x+height):
        q.append(graph[i][y+width-1])
    # 밑변 오른 -> 왼
    for i in range(y+width-2, y, -1):
        q.append(graph[x+height-1][i])
    # 왼변 아래 -> 위
    for i in range(x+height-1, x, -1):
        q.append(graph[i][y])
    q.rotate(-r)
    
    for i in range(y, y+ width):
        graph[x][i] = q.popleft()
    
    for i in range(x+1, x+height):
        graph[i][y+width-1] = q.popleft()
    
    for i in range(y+width-2, y, -1):
        graph[x+height-1][i] = q.popleft()
    
    for i in range(x+height-1, x, -1):
        graph[i][y] = q.popleft()
  
n, m, r = map(int, input().split())
graph = [list(map(int, input().split())) for _ in range(n)]

height = n
width = m
x, y = 0, 0

while True:
    if height == 0 or width == 0: 
        break
    rotate(x, y, height, width)
    x += 1
    y += 1
    height -= 2
    width -= 2
    
for i in graph:
    for j in i:
        print(j, end=" ")
    print()
```
