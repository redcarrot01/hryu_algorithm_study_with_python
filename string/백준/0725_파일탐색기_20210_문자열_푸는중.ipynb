```python
t = int(input())
arr = []*t
for _ in range(t):
    s =input()
    arr.append(s)

new_arr = [[0]*len(max(arr)) for _ in range(t)]
for i in arr:
    nums = ''
    for j in range(len(max(arr))):
        if i[j].isalpha():
            if i[j-1].isalpha():
                new_arr[i][j]=i[j]
            elif i[j-1].isdigit():
                new_arr[i][j]= nums
                nums=''
        else:
            nums += i[j]
        new_arr[i][j] = new_arr[i][j].join()    
            
        
```
