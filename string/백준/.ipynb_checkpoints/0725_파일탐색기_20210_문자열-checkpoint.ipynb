```python
import re
from functools import cmp_to_key

t = int(input())
arr = []*t
temp = []
for _ in range(t):
    s =input()
    temp = re.findall("[a-zA-Z]|\d+", s)
    arr.append([s, temp])
    
alphabet = "AaBbCcDdEeFfGgHhIiJjKkLlMmNnOoPpQqRrSsTtUuVvWwXxYyZz"

def comp(a,b):
    for i in range(min(len(a[1]), len(b[1]))):
        if a[1][i].isdigit() and b[1][i].isalpha():
            return -1
        elif a[1][i].isalpha() and b[1][i].isdigit():
            return 1
        elif a[1][i].isdigit() and b[1][i].isdigit():
            if int(a[1][i]) == int(b[1][i]):
                if len(a[1][i]) == len(b[1][i]): # 0의개수 체크
                    continue
                return len(a[1][i]) - len(b[1][i])
            else:
                return int(a[1][i]) - int(b[1][i])
        else:
            if a[1][i] == b[1][i]:
                continue
            else:
                return alphabet.index(a[1][i]) - alphabet.index(b[1][i])        
    return len(a[1]) - len(b[1])

answer = sorted(arr, key= cmp_to_key(comp))
for i in answer:
    print(i[0])
```

```python
# 문자열 나누는 연습
n = int(input())
wordlist = []
for _ in range(n):
    word_split = []
    word = input()
    digit = ""
    alpha = ""
    for character in word:
        if character.isdigit():
            digit += character
            if alpha:
                word_split.append(alpha)
                alpha = ""
        else:
            alpha += character
            print(alpha)
            if digit:
                word_split.append(digit)
                digit = ""
    if alpha:
        word_split.append(alpha)
    if digit:
        word_split.append(digit)
    wordlist.append([word, word_split])
print(wordlist)
```
