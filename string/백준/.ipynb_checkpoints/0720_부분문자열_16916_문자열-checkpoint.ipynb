### KMP 알고리즘을 이용한다

```python
def kmp():
    m = len(pat) # 패턴 길이
    n = len(txt) # 원본 길이
    lps = computeLPS()
    i ,j =0, 0
    while i < n:
        if pat[j] == txt[i]:
            i += 1
            j += 1
        elif pat[j] != txt[i]:
            if j !=0:
                j = lps[j-1]
            else:
                i += 1
        if j == m:
            return 1
    else:
        return 0
    
def computeLPS():
    leng = 0 # 접두사와 접미사가 동일한 길이
    i = 1
    lps = [0]*len(pat)
    while i< len(pat):
        if pat[i] == pat[leng]:
            leng += 1
            lps[i] = leng
            i += 1
        else:
            if leng != 0:
                leng = lps[leng -1]
            else:
                lps[i] = 0
                i += 1 
    return lps
txt = input()
pat = input()
print(kmp())
```
