```python
# 1번작업
# stack은 110을 제외한 문자열이 리스트화
# s = 10111111100  stack = 10111   cnt = 2
stack = []
cnt = 0
for c in s:
    if c =="0":
        if stack[-2:] == ['1', '1']:
            cnt+=1
            stack.pop()
            stack.pop()
        else:
            stack.append(c)
    else:
        stack.append(c)    
```

```python
# 110이 없는 경우 s가 답이됨
if cnt == 0:
    answer.append(s)
# 110 있는 경우, 끝에서부터 1 연속되는 동안은 idx 증가, 1이 연속되는부분 맨뒤에 위치하기 위해
else:
    temp = deque()
    while stack: # temp = 111, stack = 10
        if stack[-1] == '1':
            temp.append(stack.pop()) 
        elif stack[-1] =='0':
            break
```

```python
# s = 10110110111
while cnt > 0:
    temp.appendleft('0')
    temp.appendleft('1')
    temp.appendleft('1')
    cnt -= 1
while stack:
    temp.appendleft(stack.pop())
answer.append(''.join(map(str, temp)))
    
```

```python
# 전체코드
from collections import deque
def solution(s):
    answer = []
    for st in s:
        stack = []
        cnt = 0
        for c in st:
            if c == '0':
                if stack[-2: ] == ['1', '1']:
                    cnt += 1
                    stack.pop()
                    stack.pop()
                else:
                    stack.append(c)
            else:
                stack.append(c)
        
        if cnt ==0:
            answer.append(st)
        else:
            temp = deque()
            while stack:
                if stack[-1] =='1':
                    temp.append(stack.pop())
                elif stack[-1] == '0':
                    break
            
            while cnt > 0:
                temp.appendleft('0')
                temp.appendleft('1')
                temp.appendleft('1')
                cnt -= 1
                
            while stack:
                temp.appendleft(stack.pop())
            answer.append(''.join(map(str, temp)))
    return answer

solution(["1110","100111100","0111111010"])
```
