```python
import math
def solution(n):
    x = math.sqrt(n)
    if x % 1 ==0:
        answer = (x+1)**2
    else:
        answer = -1
    return int(answer)
solution(121)
```

```python
# 다른 풀이
def solution(n):
    sqrt = n**0.5
    if sqrt % 1 ==0:
        return int((sqrt+1)**2)
    return -1
solution(121)
```
