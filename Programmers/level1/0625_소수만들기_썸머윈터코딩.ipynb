```python
import itertools

def solution(nums):
    answer =0
    result = list(itertools.combinations(nums, 3))
    for i in result:
        temp = sum(i)
        prime =True
        for j in range(2, temp):
            if temp % j==0:
                prime= False
                break
        if prime==True:
            answer += 1
    return answer
    
solution([1,2,7,6,4])
```

```python
from itertools import combinations

def solution(nums):
    answer = 0
    result = list(itertools.combinations(nums,3))
    for i in result:
        temp = sum(i)
        prime = True
        for i in range(2, int(temp**0.5)+1):
            if temp %i==0:
                prime = False
                break
        if prime == True:
            answer += 1
    return answer
solution([1,2,7,6,4])   
```
