```python
def solution(a, b):
    answer = ''
    sum_days = 0
    week = ['FRI', 'SAT', 'SUN', 'MON', 'TUE', 'WED', 'THU']
    day = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    for i in range(0, a-1):
        sum_days+= day[i]
    sum_days += b-1
    answer = week[sum_days % 7]
    return answer
solution(5, 24)
solution(1,6)
```
