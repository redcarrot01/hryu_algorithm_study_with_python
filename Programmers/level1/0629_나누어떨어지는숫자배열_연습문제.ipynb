```python
def solution(arr, divisor):
    answer = []
    for i in arr:
        if i%divisor ==0:
            answer.append(i)
    answer.sort()
    if len(answer) ==0:
        answer.append(-1)
    return answer
solution([5, 9, 7, 10], 5)
```

```python
# 컴프리핸션 풀이
# 파이썬에서 or 앞이 참일경우 해당 값을 , 거짓일 경우 뒤에 것을 호출
# 빈 리스트는 무조건 false로 간주
def solution(arr, divisor):
    return sorted([n for n in arr if n%divisor ==0]) or [-1]
solution([5, 9, 7, 10], 5)
```
