```python
def solution(lottos, win_nums):
    answer = []
    lottos.sort()
    win_nums.sort()
    cnt = 0
    zero_cnt =0
    for i in range(6):
        if lottos[i]==0:
            zero_cnt += 1
        if lottos[i] in win_nums:
            cnt += 1
    
    if cnt+zero_cnt == 6:
        answer.append(1)
    elif cnt+zero_cnt == 5:
        answer.append(2)
    elif cnt+zero_cnt == 4:
        answer.append(3)
    elif cnt+zero_cnt == 3:
        answer.append(4) 
    elif cnt+zero_cnt == 2:
        answer.append(5)
    else: 
        answer.append(6)

    if cnt == 6:
        answer.append(1)
    elif cnt == 5:
        answer.append(2)
    elif cnt == 4:
        answer.append(3)
    elif cnt == 3:
        answer.append(4)
    elif cnt == 2:
        answer.append(5)
    else: 
        answer.append(6)

    return answer
solution([44, 1, 0, 0, 31, 25],[31, 10, 45, 1, 6, 19])
```

```python
# 다른 풀이
def solution(lottos, win_nums):
    # 인덱스 0, 1 -> 6위..
    rank = [6,6,5,4,3,2,1]
    
    cnt_0 = lottos.count(0)
    ans = 0
    for x in win_nums:
        if x in lottos:
            ans += 1
    return rank[cnt_0+ans], rank[ans]
solution([44, 1, 0, 0, 31, 25],[31, 10, 45, 1, 6, 19])
```
