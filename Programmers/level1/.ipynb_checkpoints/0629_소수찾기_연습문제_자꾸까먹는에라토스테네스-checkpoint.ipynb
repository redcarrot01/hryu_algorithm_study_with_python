```python
# 56점 풀이 - 실패
def solution(n):
    answer = 0
    for i in range(2, n+1):
        prime = True
        for j in range(2, i):
            if i % j ==0:
                prime =False
        if prime ==True:
            answer += 1
    return answer
solution(10)
```

```python
# 다른 풀이 - 에라토스테네스의 체
'''
에라토스테네스 : 소수를 판별한 범위만큼 배열 할당하여, 해당하는 값을 넣어주고,
이후에 하나씩 지워나가는 방법을 이용한다.
1. 배열을 생성하여 초기화
2. 2부터 시작해서 특정 수의 배수에 해당하는 수를 모두 지운다.
(지울 때 자기자신은 지우지 않고, 이미 지워진 수는 건너뛴다)
3. 2부터 시작하여 남아있는 수를 모두 출력한다.
'''
def solution(n):
    cnt = [True]*(n+1)
    answer =0
    for i in range(2, int(n**0.5)+1):
        if cnt[i]==True:
            for j in range(i+i, n+1, i):
                cnt[j]= False
    for i in range(2, n+1):
        if cnt[i]:
            answer += 1
    return answer
solution(10)
```

```python
def solution(n):
    num = set(range(2, n+1))
    for i in range(2, n+1):
        if i in num:
            num -= set(range(2*i, n+1, i))
    return len(num)
solution(10)
```
