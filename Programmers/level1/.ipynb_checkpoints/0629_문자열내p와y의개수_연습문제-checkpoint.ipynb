```python
def solution(s):
    cnt =0
    for i in s:
        if i == 'P' or i =='p':
            cnt += 1
        elif i == 'Y' or i == 'y':
            cnt -= 1
    if cnt ==0:
        return True
    else:
        return False
solution("pPoooyY")
solution("Pyy")
```

```python
# 다른 풀이
def solution(s):
    return s.lower().count('p') == s.lower().count('y')
solution("pPoooyY")
```
