```python
def solution(s):
    answer = []
    for i in s:
        if i.isupper():
            answer.append(i)
            answer.sort(reverse= True)
    temp= []
    for i in s:
        if i.islower():
            temp.append(i)
            temp.sort(reverse= True)
    temp += answer
    return "".join(temp)
solution("Zbcdefg")
```

```python
#  다른 풀이 
def solution(s):
    return ''.join(sorted(s, reverse=True))
solution("Zbcdefg")
```
