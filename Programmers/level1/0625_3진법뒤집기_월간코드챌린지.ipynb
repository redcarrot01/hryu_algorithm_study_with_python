```python
def solution(n):
    answer = 0
    temp = []
    if n<= 2:
        return n
    while n>2:
        temp.append(n%3)
        n //= 3
        if n < 3:
            temp.append(n)
    for i in range(len(temp)):
        answer += temp[i]*(3**((len(temp)-1)-i))
    return answer

solution(45)
solution(125)
```
