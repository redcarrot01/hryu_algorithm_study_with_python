## 완주하지 못한 선수

```python
# 해결 1
def solution(participant, completion):
    participant.sort()
    completion.sort()
    for i, j in zip(participant, completion):
        if i!= j :
            return i
    return participant[-1]    
```

```python
# 해결 4 : 다른 분 코드 (딕셔너리 이용)
result = {}
def solution(participant, completion):
    for a in participant:
        result[a] = result.get(a,0) + 1

    for b in completion:
        result[b] -= 1
        if result[b] == 0:
            del result[b]
    return list(result.keys())[0]

solution(["leo", "kiki", "eden"],["eden", "kiki"] )
```

```python
# 해결 2의 colletions.Counter르 이용한 카운팅
from collections import Counter

# 카운팅 1
Counter('hello world') 
# Counter({'h': 1, 'e': 1, 'l': 3, 'o': 2, ' ': 1, 'w': 1, 'r': 1, 'd': 1})

# max 카운팅 2
def find_max(world):
    counter = Counter(world)
    max_count = -1
    for letter in counter:
        if counter[letter] > max_count:
            max_count = counter[letter]
            max_letter = letter
    return max_letter, max_count
find_max('hello world') # ('l', 3)

 
# max 카운팅 3 : 데이터 개수 많은 대로
Counter('hello world').most_common()
# # [('l', 3), ('o', 2), ('h', 1), ('e', 1), (' ', 1), ('w', 1), ('r', 1), ('d', 1)]


# max 카운팅 4 : 가장 많은 값 리턴
Counter('hello world').most_common(1) # [('l', 3)]
```

## 전화번호 목록

```python
# 해결 1
def solution(phone_book):
    answer = True
    phone_book.sort()
    for i in range(len(phone_book)-1):
        if phone_book[i] in phone_book[i+1][:len(phone_book[i])]:
            return False
    return answer

solution(["119", "97674223", "1195524421"])
```

```python
# 해결 2
def solution(phone_book):
    answer =True
    hash_map = {}
    for phone_number in phone_book:
        hash_map[phone_number] =1
    for phone_number in phone_book:
        temp = ""
        for number in phone_number:
            temp += number
            if temp in  hash_map and temp != phone_number: 
            # temp가 딕셔너리에 존재하고, 접두어로 쓰이는 temp가 있다면
                return False
    return answer
solution(["119", "97674223", "1195524421"])
```

## 위장

```python
def solution(clothes):
    answer = 1
    closet = {}
    for element in clothes:
        key = element[1]
        value = element[0]
        # 딕셔너리에 키가 있다면
        if key in closet: 
            closet[key].append(value)
        else:
            closet[key] = [value]
    for key in closet.keys():
        answer = answer * (len(closet[key])+1)
    return answer -1
solution([["yellowhat", "headgear"], ["bluesunglasses", "eyewear"], ["green_turban", "headgear"]])
```

## 베스트앨범  
많이 재생된 장르 > 많이 재생된 노래 > 장르 같고, 재생 횟수 같으면 고유번호낮은거
장르 별 가장 노래 두곡만 선택함

```python
# 1. 장르별 횟수를 구한다, 2. 장르별 (곡별, 재생횟수) 구한다
def solution(genres, plays):
    answer = []
    # { 장르 : 총 재생 횟수}
    plays_dict = {}
    # { 장르 : [(재생 횟수, 고유 번호)] }
    d = {}
    # 사전 초기화
    for i in range(len(genres)):
        genre = genres[i] # 장르
        play = plays[i] # 재생 횟수
        plays_dict[genre] = plays_dict.get(genre, 0)+ play 
        d[genre] = d.get(genre, []) + [(play, i)]
        
    genre_sort = sorted(plays_dict.items(), key = lambda x:x[1], reverse=True)
    
    for (genre, total_play) in genre_sort:
        d[genre] = sorted(d[genre], key=lambda x: (-x[0], x[1]))
        answer += [i for (play, i) in d[genre][:2]]
    return answer
```
