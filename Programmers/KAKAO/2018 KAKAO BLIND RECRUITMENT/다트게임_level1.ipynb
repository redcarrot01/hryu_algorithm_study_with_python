```python
def solution(dartResult):
    dart = []
    answer = []
    bonus = ['S', 'D', 'T']
    dartResult = dartResult.replace('10', 't')
    for i in dartResult:
        if i=='t':
            i='10'
        dart.append(i)
    print(dart) #['10', 'S', '2', 'D', '*', '3', 'T']
    for i in range(len(dart)):
        if dart[i] in bonus:
            answer.append( int(dart[i-1])** (bonus.index(dart[i])+1))
        if dart[i] == '*':
            if len(answer) >= 2:
                answer[-1] = answer[-1]*2
                answer[-2] = answer[-2]*2
            else :
                answer[-1] = answer[-1]*2
        elif dart[i]=='#':
            answer[-1] = answer[-1]*(-1)
    return sum(answer)
solution('1S2D*3T')
```

```python
import re
def solution(dartResult):
    
    answer = 0
    bonus = {'S': 1, 'D': 2, 'T':3}
    option = {'*':2, '#':-1}
    
    temp = re.compile('(\d+)([SDT])([*#]?)') # ''은 객체를 리턴해줌
    dart = temp.findall(dartResult) # 매칭되는 패턴을 리스트로 반환
    #print(dart) [('1', 'S', ''), ('2', 'D', '*'), ('3', 'T', '')]
    for i in range(len(dart)):
        if dart[i][2] == '*' and i>0:
            dart[i-1] *= 2
        if dart[i][2]:
            dart[i] = int(dart[i][0])** bonus[dart[i][1]] * option[dart[i][2]]
        else:
            dart[i] = int(dart[i][0])** bonus[dart[i][1]]
    answer = sum(dart)
    return answer 
solution('1S2D*3T')
```
