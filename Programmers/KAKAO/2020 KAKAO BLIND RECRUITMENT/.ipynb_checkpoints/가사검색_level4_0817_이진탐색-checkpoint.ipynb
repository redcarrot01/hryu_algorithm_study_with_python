각 단어를 길이에 따라 나눈다. 
이후에 모든 리스트를 정렬한 뒤에, 각 쿼리에 대해 이진 탐색을 수행해야 한다.

~~~
길이 5 라수트 ["frame", "frodo", "front", "frost", "kakao"]
길이 6 리스트 ["frozen"]
~~~
'fro??' 라는 쿼리가 들어오면 'fro'로 시작하는 모든 단어를 찾는데, 이때 이진탐색을 이용하여
'fro'로 시작되는 첫 단어의 위치와 마지막 위치를 찾아서 등장한 횟수를 계산한다.

'fro??'라는 쿼리가 들어오면 count_by_range() 함수를 이용해 'froaa'보다 크거나 같으면서
'frozz'보다 작거나 같은 단어의 개수를 세도록 구현하면 된다.

접두사로 '?'가 등장하면 기존 리스트를 이용해서 처리할 수 없으므로,
기존 단어를 뒤집은 단어로 리스트를 선언해야 한다. 
~~~
['emarf', 'oakak', 'odorf', 'tnorf', 'tsorf']  
~~~
뒤집은 단어를 대상으로 이진탐색을 수행하면 된다.

```python
from bisect import bisect_left, bisect_right

# 값이 [left_value, right_value]인 데이터의 개수를 반환하는 함수
def count_by_range(a, left_value, right_value):
    right_index = bisect_right(a, right_value)
    left_index = bisect_left(a, left_value)
    return right_index - left_index

# 단어를 길이로 나눈 리스트
array = [[] for _ in range(10001)]
reversed_array = [[] for _ in range(10001)]

def solution(words, queries):
    answer = []
    # 모든 단어를 접미사 와일드카드 배열, 접두사 와일드카드 배열에 각각 삽입
    for word in words:
        array[len(word)].append(word) # 단어를 삽입
        reversed_array[len(word)].append(word[::-1]) # 단어를 뒤집어서 삽입
    print(array)
    print(reversed_array)
    # 이진 탐색 수행위한 단어 리스트 정렬 수행
    for i in range(10001):
        array[i].sort()
        reversed_array[i].sort()
        
    for q in queries:
        if q[0] != "?":
            res = count_by_range(array[len(q)],q.replace('?', 'a'), q.replace('?', 'z') )
        else:
            res = count_by_range(reversed_array[len(q)], q[::-1].replace('?', 'a'), q[::-1].replace('?', 'z'))
        answer.append(res)
    return answer
```
