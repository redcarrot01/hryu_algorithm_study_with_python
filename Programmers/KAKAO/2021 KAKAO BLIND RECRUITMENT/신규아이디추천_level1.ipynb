### 패턴 객체의 메서드 -  findall()
정규식에 매치되는 모든 문자열을 리스트 형식으로 리턴한다.  

### strip 
공백 제거 , 변수.strip() 

https://nachwon.github.io/regular-expressions/

```python
import re
def solution(new_id):
    # 1
    new_id = new_id.lower()
    # 2 - 정규표현식 사용
    new_id = ''.join(re.findall("[a-z0-9-._]+", new_id))
    # 2 - 원소 하나하나 확인
#     result = ''
#     for i in new_id:
#         if i.isalnum() or i in '-._':
#             result += i
    
    # 3 - 정규표현식 사용
    #new_id = re.sub('[..]+', '.', new_id)
    # 3 - while 사용
    while '..' in new_id:
        new_id = new_id.replace('..', '.')
    
    # 4
    new_id = new_id.strip('.')
    
    # 5 
    if len(new_id) ==0:
        new_id += 'a'
    
    # 6
    if len(new_id) >= 16: 
        new_id = new_id[:15]
        new_id = new_id.strip('.')
        
    # 7
    while len(new_id) <= 2:
        if len(new_id) == 3:
            break
        new_id += new_id[-1]
        
    return new_id
solution("...!@BaT#*..y.abcdefghijklm")
```

```python
# 다른 풀이
import re
def solution(new_id):
    # 1
    new_id = new_id.lower()
    # 2
    new_id = re.sub('[^a-z0-9-._]', '', new_id) #[^] 제외한다
    # 3
    new_id = re.sub('\.+', '.', new_id) #'\.' 마침표를 가리킬때 \ 필요 
    # 4 
    new_id = re.sub('^[.]|[.]$', '', new_id) # '^' 시작
    # 5
    new_id = 'a' if len(new_id)==0 else new_id[:15]
    # 6
    new_id = re.sub('^[.]|[.]$', '', new_id)
    # 7
    while len(new_id) < 3: new_id += new_id[-1]
    
    return new_id
solution("...!@BaT#*..y.abcdefghijklm")
```
