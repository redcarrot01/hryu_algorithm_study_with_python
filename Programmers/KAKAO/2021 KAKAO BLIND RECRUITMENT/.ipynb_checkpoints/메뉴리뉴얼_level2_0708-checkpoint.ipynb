리스트, Counter, 딕셔너리를 적용하여 문제를 해결했다.   
문제에서 제시한 대로 구현하면 되는데, 이때 조건을 잘 살피고 예외 처리를 잘 해줘야 한다.  

 1. **course(구성하고자 하는 메뉴 개수)를 기준으로 반복문을 돌려서 메뉴의 조합을 찾는다.**  
	- orders(메뉴들)에서 하나씩 꺼내서 메뉴 개수를 기준으로 정렬된 조합을 구하고, temp 리스트에 추가한다.  
    - temp에서 각 원소에 대해 개수를 세고, 내림차순으로 정렬한다. (Counter를 적용하면 된다)  
    - Counter는 개수 내림차순으로 원소:개수 정렬하는 방법이 있다. -> `tmp = result.most_common()`   
    - print(tmp) ->  `# Counter({('A', 'C'): 4...에서 [(('A', 'C'), 4)..로 바뀜`  

   <br>
    2. **가장 많이 주문한 단품메뉴세트를 점검 해야 한다**.  
    (그 이유는 손님들로부터 최소 2명이상에게 선택받아야 한다는 조건과 단품메뉴가 구성되지 않는 경우를 예외 처리 해줘야 한다.)     
    - 가장 주문한 메뉴 횟수를 maximum으로 설정하고, 이것이 2회 이상이고, maximum과 동일한 횟수의 메뉴가 있다면 answer에 추가한다.     

```python
from itertools import combinations
from collections import Counter
def solution(orders, course):
    answer = []
    for c in course:
        temp = []
        for order in  orders:
            combi_list = combinations(sorted(order), c)
            temp += combi_list
        #print(temp)
        result = Counter(temp) # Counter는 개수 내림차순으로 정렬된다
        tmp = result.most_common() # Counter({('A', 'C'): 4...에서 [(('A', 'C'), 4)..로 바뀜
        #print(tmp)
        
        if tmp:
            maximum = tmp[0][1]
            if maximum >= 2:
                for key, value in result.items():
                    if result[key] == maximum:
                        answer.append(''.join(key))
                        
    return sorted(answer)
#solution(["ABCFG", "AC", "CDE", "ACDE", "BCFG", "ACDEH"],[2,3,4])
#solution(["ABCDE", "AB", "CD", "ADE", "XYZ", "XYZ", "ACD"],[2,3,5])
solution(["XYZ", "XWY", "WXA"],[2,3,4])
```

```python
Counter 대신에 딕셔너리를 사용하여 메뉴의 개수를 세고, sort를 이용해 개수를 내림차순으로 정렬했다.
```

```python
from itertools import combinations
def solution(orders, course):
    answer = []
    for c in course:
        dic = {}
        for order in orders:
            combi_list = combinations(sorted(order), c)
            for i in combi_list:
                if i in dic:
                    dic[i] += 1
                else:
                    dic[i] = 1
        dic = sorted(dic.items(), reverse=True, key =lambda x:x[1])
        for i in range(len(dic)):
            if dic[i][1] == dic[0][1] :
                if dic[i][1] >= 2:
                    answer.append(''.join(dic[i][0]))
    return sorted(answer)
solution(["XYZ", "XWY", "WXA"],[2,3,4])
```
