```python
def solution(n):
    dp= [0]* 100001
    dp[1] = 1
    dp[2] = 1
    for i in range(3, n+1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[i] % 1234567
solution(5)
```

```python
# 런타임 에러 발생
d = [0] * 100001
def solution(n):
    # 종료 조건
    if n ==1 or n==2:
        return 1
    # 메모이제이션
    if d[n] !=0:
        return d[n]
    d[n] = solution(n-1) + solution(n-2)
    return d[n] % 1234567
solution(5)
```
