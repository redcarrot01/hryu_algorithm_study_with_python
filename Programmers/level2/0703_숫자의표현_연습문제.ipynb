```python
# 1부터 n까지 순회하면서 더해주는데, n이랑 같으면  카운트하고 break, n보다 크다면 안쪽 반복문 break 
```

```python
def solution(n):
    answer = 0
    for i in range(1, n+1):
        result =0
        for j in range(i, n+1):
            result += j
            if result ==n:
                answer += 1
                break
            if result > n:
                break
    return answer
```
