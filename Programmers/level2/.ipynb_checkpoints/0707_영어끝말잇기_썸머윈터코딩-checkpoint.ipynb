```python
# 끝말잇기가 성립 안되거나, 증복된 단어 있는 경우 
def tail_and_head(words):
    for i in range(1, len(words)):
        if words[i-1][-1] != words[i][0] or words[i] in words[:i]:
            return i
    return 0

def solution(n, words):
    answer = [0,0]
    #  끝말잇기 가능하고, 중복된 단어 없는 경우
    if tail_and_head(words)==0:
        return answer
    else:
        result = tail_and_head(words)
#         print(result)
        if (result+1)%n == 0:
            answer[0] = n
            answer[1]= (result+1)//n
        else:
            answer[0]= (result+1)%n
            answer[1]= (result+1)//n +1
        return answer

solution(3,["tank", "kick", "know", "wheel", "land", "dream", "mother", "robot", "tank"])
# solution(5,	["hello", "observe", "effect", "take", "either", "recognize", "encourage", "ensure", "establish", "hang", "gather", "refer", "reference", "estimate", "executive"] )
solution(2,["hello", "one", "even", "never", "now", "world", "draw"] )
```

```python
# 다른 풀이
def solution(n, words):
    answer = []
    turn = 0
    wordList = [words[0]]
    # [실행] 버튼을 누르면 출력 값을 볼 수 있습니다.
    for idx in range(1, len(words)):
        if words[idx-1][-1] != words[idx][0]:
            turn = idx
            break
        if words[idx] in wordList:
            turn = idx
            break
        wordList.append(words[idx])
    answer = [turn%n +1, turn//n + 1]
    if turn == 0:
        answer = [0, 0]
    return answer
```
