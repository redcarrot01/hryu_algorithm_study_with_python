```python
1. 문자열을 리스트로 바꾼다.
2. 문자열 리스트를 정수형 리스트로 바꾼다.
3. 최솟값과 최댓값을 구하고, 이것을 문자열로 바꾼다.
```

```python
def solution(s):
    answer = ''
#     s_list = s.split(" ")
#     s_list = list(map(int, s_list))
    s_list = list(map(int, s.split(' ')))
    answer = str(min(s_list))+" "+str(max(s_list))
    return answer
solution("-1 -2 -3 5")
```
