```python
def grade(sum_):
    ans =''
    if 90<= sum_:
        ans += 'A' 
    elif 80 <= sum_ < 90:
        ans += 'B'
    elif 70 <= sum_ < 80:
        ans += 'C'
    elif 50 <= sum_ < 70:
        ans += 'D' 
    else:
        ans+= 'F'
    return ans

def solution(scores):
    answer = ''
    average = []
    n= len(scores)
    for j in range(n):
        sum_ = 0
        for i in range(n):
            average.append(scores[i][j])     
        if (average[j] == max(average) or  average[j] == min(average)) and average.count(average[j])==1:
            del(average[j])
        answer += grade(sum(average)/len(average))
        average = []
            
    return answer
```
