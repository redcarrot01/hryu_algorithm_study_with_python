```python
# 시간 초과 나는 풀이 
n, m = map(int, input().split())
arr = list(map(int, input().split()))

sum_value = 0
sum_arr = [0]
cnt = 0
for i in arr:
    sum_value += i
    sum_arr.append(sum_value)
#print(sum_arr)
for i in range(1, n+1):
    for j in range(0, i):
        if (sum_arr[i]- sum_arr[j]) ==m:
            print(sum_arr[i], sum_arr[j])
            cnt += 1
print(cnt)
```

```python
from collections import defaultdict
n, m = map(int, input().split())
arr = list(map(int, input().split()))

cnt = 0
dic = defaultdict(int)
for i in range(1, len(arr)):
    arr[i] += arr[i-1]
    
for i in range(len(arr)):
    if arr[i] ==m:
        cnt += 1
    cnt += dic[arr[i]-m]
    dic[arr[i]] += 1
print(cnt)
```
