```python
# 1이 될 때 까지

n, k = map(int, input().split())
result = 0

while n>= k:
    while n % k != 0:
        n-=1
        result += 1
    n //= k
    result += 1
    
while n >1:
    n-= 1
    result += 1
print(result)
```

```python
# 문자열 뒤집기
data = input()

cnt0 = 0
cnt1 = 0

if data[0] == '0':
    cnt1 += 1
else:
    cnt0 += 1

for i in range(len(data)-1):
    if data[i] != data[i+1]:
        if data[i+1] == '0':
            cnt1+= 1
        else:
            cnt0 += 1
            
print(min(cnt0, cnt1))
```

```python
# 더하기 혹은 곱하기
# 1이하이면 +, 다른 경우 *

data = input()
result = int(data[0])
for i in range(1, len(data)):
    num = int(data[i])
    if num <= 1 or result <= 1:
        result += num
    else:
        result *= num
print(result)
```
