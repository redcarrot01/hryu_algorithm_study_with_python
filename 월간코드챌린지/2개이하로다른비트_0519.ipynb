```python
# 81.8점 
'''주어진 숫자를 이진수로 바꿔서 바꾼 수를 오름차순으로 비교하여
조건에 부합하면 리턴해주도록 구현했지만 -> 시간초과 발생
'''
def solution(numbers):
    answer = []
    for i in numbers:
        b_i= bin(i)
        temp=i+1
        while True:
            b_temp = bin(temp)
            s_i = str(b_i)
            s_i = s_i.replace("b","0")
            
            s_temp = str(b_temp)
            s_temp = s_temp.replace("b","0")
            cnt = 0

            s_i_r = s_i[::-1]
            s_temp_r = s_temp[::-1]
            
            for t in range(len(s_i_r)):
                if s_i_r[t] != s_temp_r[t]:
                    cnt += 1
                    if cnt >2:
                        break
                        
            if  0<cnt<= 2: 
                answer.append(temp)
                break
            else:
                temp += 1
    return answer
solution([2,7])
```

비트연산을 이용해야 한다

경우 1 짝수
'2' = 0010 → f(2) = '3' = 0011
'4' = 0100 → f(4) = '5' = 0101
'6' = 0110 → f(6) = '7' = 0111
'8' = 1000 → f(8) = '9' = 1001
'10' = 1010 → f(10) = '11' = 1011
1. 주어진 숫자들이 모두 짝수
2. 주어진 숫자들의 f(x) 값은 모두 x + 1
3. 짝수 x는 최하위 비트가 0이고, 최하위 비트를 1로 바꾸면 x+1이되며, 이때 x보다 크면서 가장 작은 숫자 만족하는 서로 다른 비트 갯수 1개


경우2 홀수
'7' = 0111 → f(7) = 11 = 1011
'9' = 1001 → f(9) = 10 = 1010
'11' = 1011 → f(11) = 13 = 1101
1. 주어진 숫자가 홀수
2. 0중에서 가장 최하위 비트에 있는 0을 1로 바꾸고, 그다음비트를 0으로 바꿔준다

구현

예) 
1001이 있다. 
1. 최하위 비트 0의 연산을 위해 1로 변환
최하위 비트에 존재하는 0을 1로 바꿔주고 연산할 것
1001 + 1 = 1010 (최하위 0인 비트가 1로 바뀜)

2. 2의 보수 구하기(구하는법 : 비트 반전 + 1)
1001의 2의 보수는 1001반전 -> 0110 , 0110+1 -> 0111

3. 1의 변환한 수와 2의보수를 and 연산하면 최하위 비트0 위치 찾음
최종적으로 1010 & 0111 -> 0010 , 최하위 0인 비트만 1로 표시됨(최하위 비트 0 위치 찾음)

4. 최하위비트 0을 1로 바꾸기
1001 | 0010(0의 위치) -> 1011

5. 최하위비트 0의 다음 비트1을 0으로 바꿔주기
최하위비트를 >> 1한곳에 위치한 비트를 바꿔줄 것, 동시 나머지 비트들은 본래 유지
0010 >> 1 : 비트를 오른쪽으로 1번 이동 0001
0001을 반전 1110 (반전하는 이유는 1의비트를 0으로 바꿔주는 것이 목표니까)
1011(최하위비트 0->1로 바꾼 비트) & 1110 -> 1010(우리가 구하고자 하는 )


```python
# https://yabmoons.tistory.com/665 참고한 풀이
def solution(numbers):
    answer = []
    for num in numbers:
        if num %2 ==0:
            answer.append(num+1)
        else:
            # 1,2,3 ->  최하위 비트 0 위치 찾기  
            lastZero = (num+1) & (-num)
            # 4 -> 최하위 비트 0을 1로 바꾸기
            zero_toOne = num | lastZero  
            # 5 -> 최하위 비트 0의 다음 비트1을 0으로 변환
            temp = (zero_toOne & (~(lastZero)>> 1))
            answer.append(temp)
    return answer
solution([2,7])                    
      
```

```python
# 다른 풀이 -> 내풀이

def solution(numbers):
    answer = []

    for num in numbers:
        bin_num = bin(num)
        temp = '0' + bin_num[2:] 
        
        lastZero = -1
        for i in range(len(temp)):
            if temp[i] == '0':
                lastZero = max(lastZero, i)

        if num%2 ==0:
            answer.append(num+1)
        else:
            result = temp[:lastZero] + '10' + temp[lastZero+2:]
            answer.append(int(result,2))
    return answer
solution([2,7])
```
