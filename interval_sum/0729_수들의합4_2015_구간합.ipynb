```python
# 시간 초과 나는 풀이 
n, m = map(int, input().split())
arr = list(map(int, input().split()))

sum_value = 0
sum_arr = [0]
cnt = 0
for i in arr:
    sum_value += i
    sum_arr.append(sum_value)
#print(sum_arr)
for i in range(1, n+1):
    for j in range(0, i):
        if (sum_arr[i]- sum_arr[j]) ==m:
            print(sum_arr[i], sum_arr[j])
            cnt += 1
print(cnt)
```

```python
n, m = map(int, input().split())
arr = list(map(int, input().split()))

sum_value = 0
cnt = 0
mod = [0]*m
for i in range(len(arr)):
    sum_value += arr[i]
    temp = sum_value%m
    if temp ==0:
        cnt += 1
    mod[temp] += 1
for i in range(len(mod)):
    cnt += mod[i]*(mod[i]-1)//2
print(cnt)
```
