```python
# 대실패 
    def dfs(x, y, temp):
    dx = [-1, 1, 0 ,0]
    dy = [0, 0, -1, 1]
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        
        if visited[nx][ny] == 0 and nx >=0 and nx<n and ny>=0 and ny<m :
            visited[nx][ny] =1
            if temp <= temp+graph[nx][ny]:
                temp += graph[nx][ny]
                dfs(nx, ny, temp)
                return temp
            else:
                break
            #visited[nx][ny] =1
            
                

n,m = map(int, input().split())
graph = [list(map(int, input().split())) for _ in range(n)]
visited = [[0]*m for _ in range(n)]

result = int(-1e9)
for i in range(n):
    for j in range(m):
        if visited[i][j] == 0:
            visited[i][j] =1
            temp = dfs(i, j, graph[i][j])
            #temp = max(temp, temp+graph[0][0])
            result= max(result, temp)
            temp = 0
print(result)

```
