```python
이 문제는 모든 노드를 탐색해야 하므로 완전탐색, 안전 영역을 계산할 때는 dfs를 적용했다.

풀이 순서
3개의 벽을 만들고, dfs를 통해 바이러스를 퍼뜨린다. 안전 영역의 크기를 세고, 최댓값을 찾는다.
원본 맵을 잃지 않기 위해 맵을 복사하는 과정이 필요하다. 복사된 맵은 3개의 벽이 새로 세워진 맵이 복사된다.
1. 3개의 벽을 세우는 과정은 최악의 경우 64개의 공간 중 3개를 택하는 조합의 경우이고,
이것은 100,000보다 작은 수이므로, 완전 탐색을 진행해도 시간 안에 문제 해결이 가능하다.
2. 바이러스를 퍼뜨리는 경우는 DFS의 재귀를 사용하여 구현했다.
```

```python
n, m = map(int, input().split())
data = []
temp = [[0]*m for _ in range(n)]

for _ in range(n):
    data.append(list(map(int, input().split())))
dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

result =0

# 바이러스 퍼뜨리기
def virus(x, y):
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        # 상 하 좌 우 중에서 바이러스 퍼질 경우
        if 0<=nx<n and 0<=ny<m :
            if temp[nx][ny] ==0:
                # 해당 위치에 바이러스 배치
                temp[nx][ny] = 2
                virus(nx, ny)

# 안전 영역 계산               
def get_space():
    space =0
    for i in range(n):
        for j in range(m):
            if temp[i][j] == 0:
                space += 1
    return space
  
# DFS 이용해 각 바이러스 퍼뜨리기
def dfs(cnt):
    global result
    if cnt ==3:
        for i in range(n):
            for j in range(m):
                temp[i][j] = data[i][j]
        
        for i in range(n):
            for j in range(m):
                if temp[i][j] == 2:
                    virus(i,j)
        result = max(result, get_space())
        return 
    
    # 빈 공간에 울타리 설치
    for i in range(n):
        for j in range(m):
            if data[i][j] == 0:
                data[i][j] =1
                cnt += 1
                dfs(cnt)
                data[i][j] =0
                cnt -= 1

dfs(0)
print(result)
    
```
