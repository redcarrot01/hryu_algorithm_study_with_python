```python
1. 장애물, 교사, 학생의 좌표를 각각 리스트에 저장
2. 장애물의 조합을 구하고, 장애물을 설치한 뒤, 교사의 이동 가능한 좌표 체크 
3. 교사는 상하좌우 네 방향에 대해 장애물이 없다면 학생 좌표까지 포함하여 영역을 확장 - dfs함수
4. visit함수를 통해 교사의 영역과 학생의 영역 중 겹치는 부분이 있다면 false/ 아니라면 true 반환
5. 메인에서 visit이 참으로 리턴되는 경우가 있다면, 감시피하기는 성공이므로 true를 리턴하고 메인 탈출
```

```python
from itertools import combinations
from copy import deepcopy

n = int(input())
graph = [list(map(str, input().split())) for _ in range(n)]

temp = [] # 장애물을 세울 수 있는 좌표 리스트
teachers = [] 
students = []

for i in range(n):
    for j in range(n):
        if graph[i][j] == 'X':
            temp.append([i, j])
        elif graph[i][j] == 'T':
            teachers.append([i, j])
        elif graph[i][j] == 'S':
            students.append([i, j])
            
dx = [1, 0, -1, 0]
dy = [0, 1, 0, -1]

# 장애물을 세웠다면,
def visit():
    copy_graph = deepcopy(graph) # 원본 보호하는 복사맵
    for [x,y] in teachers:
        for i in range(4):
            dfs(copy_graph, x, y, i)
    for [x,y] in students:
        if copy_graph[x][y] != 'S':
            return False
    return True

            
def dfs(graph, x, y, i):
    global n
    if x < 0 or x >= n or y < 0 or y >= n or graph[x][y] == 'O':
        return
    else:
        graph[x][y] = 'T'
        dfs(graph, x+dx[i], y+dy[i], i)  
 

check = False
temp_list =list(combinations(temp, 3))
for i in temp_list:
    for [x,y] in i:
        graph[x][y] = 'O'
    if visit():
        check=True
        break
    for [x,y] in i:
        graph[x][y] = 'X' # 복구
if check:
    print("YES")
else:
    print("NO")
```
