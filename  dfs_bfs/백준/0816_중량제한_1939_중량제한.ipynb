# bfs + 파라메트릭 서치

최댓값을 찾는 문제로, 답이 여러개가 될 수 있으며 그중 최적화된 답을 찾아야 하는데, 
이런 최적화 문제는 결정문제로 바꿔서 푸는 파라메트릭 서치로 해결할 수 있다.

이분 탐색은 구하고자 하는 값, TARGET을 찾는다면 파라메트릭 서치는 최댓값을 찾음으로써 최적화된 답을 도출한다.
파라메트릭 서치는 ‘가장 좋은 답은 무엇인가?’ 라는 최적화 문제를 ‘x 혹은 그보다 좋은 답이 있는가? YES / NO' 라는 결정 문제 형태로 바꿔서 해결하는데, 이때 이분탐색을 이용한다.
이분탐색은 배열에서 X와 일치하는 값을 찾으면 위치를 리턴하고 종료하지만, 파라메트릭 서치는 종료하지 않고 더 살펴볼 배열이 남아있는지에 대해 탐색을 계속 수행한다.

# 크루스칼 알고리즘
최대한 무거운 것을 옮기기 위한 다리만 선정한 경로를 만들어야하므로 최대 스패닝 트리를 구현한다.
중량 제한이 큰 정점들을 선택하면서 스패닝 트리를 만들어가면, 어떤 정점 S와 다른 정점 E이 같은 "연결된 집합"에 속하는 그 순간에 우리가 원하는 경로가 완성된다.

크루스칼은 bfs보다 훨씬 빠르다. 탐색 하지 않고 최적의 값을 바로 찾아낸다.

```python
import sys
input = sys.stdin.readline

# 특정 원소가 속한 집합 찾기
def find(edges, x):
    if edges[x] != x:
        # 루트노드 찾을때까지 재귀호출
        edges[x] = find(edges, edges[x])
    return edges[x]

# 두 원소가 속한 집합 찾기
def union(edges, a, b):
    a = find(edges, a)
    b = find(edges, b)
    if a<b:
        edges[b] = a
    else:
        edges[a] = b

n, m = map(int,input().split())
graph = []
edges = [i for i in range(n+1)]
for i in range(m):
    a, b, c = map(int, input().split())
    graph.append([c, a, b])
d, e = map(int, input().split())

graph.sort(reverse=True)
for i in graph:
    c, a, b = i[0], i[1], i[2]
    union(edges,a , b)
    if find(edges,d) == find(edges,e):
        print(c)
        break    
```
