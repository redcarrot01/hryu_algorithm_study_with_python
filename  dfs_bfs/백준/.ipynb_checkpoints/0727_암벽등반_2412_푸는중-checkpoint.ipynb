```python
from collections import deque
n, t = map(int, input().split())
graph = [list(map(int, input().split())) for _ in range(n)]
#graph = sorted(graph, key= lambda x:x[1])
graph.sort()
print(graph)
q= deque()
q.append([0, 0])
cnt =0
visit =[]
while q:
    x,y =q.popleft()
    for i in range(len(graph)-1):
        if y == t:
            cnt += 1
            break
        if [x,y] not in visit and abs(graph[i][0]-x) <= 2 and abs(graph[i][1]-y) <=2 :
            if abs(graph[i+1][0]-x) >2 or abs(graph[i+1][1]-y) >2:
                visit.append([x,y])
                cnt += 1
                q.append((graph[i][0], graph[i][1]))
                print(graph[i][0], graph[i][1])
if cnt ==n:
    print(-1)
else:
    print(cnt+1)           
            
```
