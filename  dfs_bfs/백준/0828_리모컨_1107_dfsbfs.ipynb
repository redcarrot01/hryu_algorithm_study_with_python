```python
완전탐색으로 최소 횟수를 구한다.
이동하려는 채널로 가는 방법은
1. 앞뒤로 +1/-1
2. 고장나지 않은 채널이라면 바로 이동
시작 번호는 100
```

```python
# 주석 코드
#import sys
#input = sys.stdin.readline
s = int(input())
n = int(input())
err = list(map(str, input().split()))
cnt = abs(100 - s) # 100 +/- 이동하는 경우
for i in range(1000001): # 50000보다 큰 수에서 작은 수로 이동하는 경우를 포함 
    i = str(i) # i=[1, 2, 3]인 경우라 가정 
    for j in range(len(i)):
        if i[j] in err: # err = [5, 6, 7] 가정, i에 있는 문자는 err 채널 아님
            break
        elif j == len(i)-1: # j가 순회하면서 i의 마지막 채널까지 왔다면 cnt 비교
            
            # 해당 번호로부터 타겟까지의 차이 + 번호를 누른 횟수와 비교
            cnt = min(cnt, abs(s-int(i)) + len(str(i)) )
print(cnt)
```
