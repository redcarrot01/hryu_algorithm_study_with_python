## 체육복

```python
# 해결1
def solution(n, lost, reserve):
    # 차집합 하기 위해 set으로 변경
    reserve_set = set(reserve)-set(lost)
    lost_set = set(lost)-set(reserve)
    
    # reserve를 기준으로 반복한다. reserve는 한번 탐색한 것은 건너뛴다.
    # 여별을 한번에 두명에게 빌려주지 않기 위함이다.
    for i in list(reserve_set):
        if i-1 in lost_set:
            lost_set.remove(i-1)
            
        elif i+1 in lost_set:
            lost_set.remove(i+1)
    return n-len(lost_set)

solution(5, [2,4], [1,3,5])
```

```python
# 해결 2
def solution(n, lost, reserve):
    reserve_list = [i for i in reserve if i not in lost]
    lost_list = [j for j in lost if j not in reserve]
    
    for i in reserve:
        a = i-1
        b = i+1
        if a in lost_list:
            lost_list.remove(a)
        elif b in lost_list:
            lost_list.remove(b)
    return n-len(lost_list)
solution(5, [2,4], [1,3,5])
```

```python
''' 
1. 안 잃어버린 친구들 카운트, 잃어버렸지만 여분 있는 친구도 추가
2. 잃어버리고 여분 없는 친구들 중에서 빌리는게 가능하면 카운트, 여분제거해준다.  
'''
def solution(n, lost, reserve):
    answer = 0
    for i in range(1, n+1):
        if i not in lost: # 안 잃어버린 학생
            answer += 1
        else:
            if i in reserve: # 잃어버렸지만 여분도 있는 학생
                answer += 1
                reserve.remove(i)
    # 잃어버리고, 여분 없는 학생
    for i in lost:
        if i-1 in reserve: 
            answer +=1
            reserve.remove(i-1)
        elif i+1 in reserve:
            answer += 1
            reserve.remove(i+1)
    return answer
solution(5, [2,4], [1,3,5])
```

## 조이스틱

```python
def solution(name):
    answer =0
    joystick = [min(ord(i)-ord('A'),ord('Z')- ord(i) + 1) for i in name ]
    idx =0
    
    while True:
        answer += joystick[idx]
        joystick[idx] = 0
        if sum(joystick) == 0:
            return answer
        left, right = 1, 1
        while joystick[idx-left]==0:
            left += 1
        while joystick[idx+right]==0:
            right += 1
        answer += min(left, right)
        idx += -left if left<right else right
    return answer

solution("JEROEN")
```
