## 모의고사

```python
def solution(answers):
    answer = []
    n1 = [1, 2, 3, 4, 5]
    n2 = [2, 1, 2, 3, 2, 4, 2, 5]
    n3 = [3, 3, 1, 1, 2, 2, 4, 4, 5, 5]
    c1 =c2= c3 =0
    for i in range(len(answers)):
        if answers[i] == n1[i%5]:
            c1 += 1
        if answers[i] == n2[i%8]:
            c2 += 1
        if answers[i] == n3[i%10]:
            c3 += 1
    maxd = max(c1, c2, c3)

    if maxd == c1: answer.append(1)
    if maxd == c2: answer.append(2)
    if maxd == c3: answer.append(3)
    
    return answer
solution([1, 2, 3, 4, 5])
solution([1, 3, 2, 4, 2])
```

## 소수찾기

```python
# 17 -> 7 17 71
# 011 -> 11, 101
from itertools import permutations
import math

def check(n):
    k = math.sqrt(n)
    if n<2:
        return False
    for i in range(2, int(k)+1):
        if n%i ==0:
            return False
    return True
    
def solution(numbers):
    answer = []
    for k in range(1, len(numbers)+1):
        perlist = list(map(''.join, permutations(list(numbers),k)))
        #print(perlist) #['1', '7']  , ['17', '71']
        
        for i in list(set(perlist)):
            if check(int(i)):
                answer.append(int(i))
    
    answer = len(set(answer)) # 앞에가 0이 있는 011, 11 -> 둘다 같은 수니까
    return answer
solution("011")
```

```python
### 에라토스테네스 체
def prime_list(n):
    # 에라토스테네스의 체 초기화 : n개 요소에 True 설정(소수로 간주)
    numbers = [True] * n
    
    # n개 최대 약수가 squrt(n) 이하이므로 i=sqrt(n)까지 검사
    m = int(n**0.5)
    for i in range(2, m+1):
        if numbers[i] ==True: # i가 소수인경우
            for j in range(i+i, n, i): # i 이후 i 배수들을 False로 
                numbers[j] = False
    return [i for i in range(2, n) if numbers[i]==True]

prime_list(10)
```
