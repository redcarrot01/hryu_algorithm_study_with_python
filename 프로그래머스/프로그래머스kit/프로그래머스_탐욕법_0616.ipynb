## 체육복

```python
# 해결1
def solution(n, lost, reserve):
    # 차집합 하기 위해 set으로 변경
    reserve_set = set(reserve)-set(lost)
    lost_set = set(lost)-set(reserve)
    
    # reserve를 기준으로 반복한다. reserve는 한번 탐색한 것은 건너뛴다.
    # 여별을 한번에 두명에게 빌려주지 않기 위함이다.
    for i in list(reserve_set):
        if i-1 in lost_set:
            lost_set.remove(i-1)
            
        elif i+1 in lost_set:
            lost_set.remove(i+1)
    return n-len(lost_set)

solution(5, [2,4], [1,3,5])
```

```python
# 해결 2
def solution(n, lost, reserve):
    reserve_list = [i for i in reserve if i not in lost]
    lost_list = [j for j in lost if j not in reserve]
    
    for i in reserve:
        a = i-1
        b = i+1
        if a in lost_list:
            lost_list.remove(a)
        elif b in lost_list:
            lost_list.remove(b)
    return n-len(lost_list)
solution(5, [2,4], [1,3,5])
```

```python
''' 
1. 안 잃어버린 친구들 카운트, 잃어버렸지만 여분 있는 친구도 추가
2. 잃어버리고 여분 없는 친구들 중에서 빌리는게 가능하면 카운트, 여분제거해준다.  
'''
def solution(n, lost, reserve):
    answer = 0
    for i in range(1, n+1):
        if i not in lost: # 안 잃어버린 학생
            answer += 1
        else:
            if i in reserve: # 잃어버렸지만 여분도 있는 학생
                answer += 1
                reserve.remove(i)
    # 잃어버리고, 여분 없는 학생
    for i in lost:
        if i-1 in reserve: 
            answer +=1
            reserve.remove(i-1)
        elif i+1 in reserve:
            answer += 1
            reserve.remove(i+1)
    return answer
solution(5, [2,4], [1,3,5])
```

## 조이스틱

```python
def solution(name):
    answer =0
    joystick = [min(ord(i)-ord('A'),ord('Z')- ord(i) + 1) for i in name ]
    idx =0
    
    while True:
        answer += joystick[idx]
        joystick[idx] = 0
        if sum(joystick) == 0:
            return answer
        left, right = 1, 1
        while joystick[idx-left]==0:
            left += 1
        while joystick[idx+right]==0:
            right += 1
        answer += min(left, right)
        idx += -left if left<right else right
    return answer

solution("JEROEN")
```

```python
# 시간초과 나는 풀이1
from itertools import combinations

def solution(number, k):
    answer = ''
    number = list(number)
    num_list = list(combinations(number,len(number)-k))
    answer = "".join(max(num_list))
    
    return answer
solution("1231234",3)
```

```python
def solution(number, k):
    l = len(number)-k
    stack = []
    for i in range(len(number)):
        if not stack:
            stack.append(number[i])
        else:
            if stack[-1] >= number[i]:
                stack.append(number[i])
            else:
                while k>0 and len(stack)>0 and stack[-1] < number[i]:
                    stack.pop()
                    k -= 1
                stack.append(number[i])

    stack=stack[:l]
    return ''.join(stack)

solution("1231234",3)
```

## 구명보트

```python
'''
*이 문제의 포인트: 가장 무거운 사람을 태우되 남은 1자리를 
가벼운 사람부터 차례대로 고려한다. 남은 사람들 중 가장 가벼운데 
무게가 초과됬다는 말은 두번째로 가벼운 사람 역시 초과될 것이라는 뜻. 
더 이상 고려할 필요가 없게 된다. 초과되면 무거운 사람만 보트에 
한명 태우고 나머지 형식도 반복해서 최대한 무거운+가벼운 사람의 조합을 
사용해 보트를 최소화 한다.
'''
```

```python
def solution(people, limit):
    answer = 0
    people.sort()
    right = len(people)-1
    i = 0
    while i<= right:
        if people[i]+people[right] > limit:
            answer += 1
            right -= 1
        else:
            i+=1
            right-=1
            answer += 1
    return answer
    
    
solution([70, 50, 80, 50], 100)
```

## 섬 연결하기

```python
# 섬과 섬 연결하기
def union(parent, a, b):
    A = find(parent, a)
    B = find(parent, b)
    if A < B:
        parent[B] = A
    else:
        parent[A] = B


def find(parent, x):
    if parent[x] == x:
        return x
    return find(parent, parent[x])
    
    
def solution(n, costs):
    parents = [i for i in range(n)]
    costs = sorted(costs, key = lambda x: x[2])
    answer = 0
    while costs:
        city1, city2, cost = costs.pop(0)
        if find(parents, city1) != find(parents, city2):
            union(parents, city1, city2)
            answer += cost
    return answer
solution(4,[[0,1,1],[0,2,2],[1,2,5],[1,3,1],[2,3,8]])
```

## 단속카메라

```python
def solution(routes):
    routes.sort(key=lambda x : x[1])
    cnt =0
    camera = -30001
    
    for route in routes:
        if camera < route[0]:
            cnt += 1
            camera = route[1]
    return cnt
solution([[-20,15], [-14,-5], [-18,-13], [-5,-3]])
```
