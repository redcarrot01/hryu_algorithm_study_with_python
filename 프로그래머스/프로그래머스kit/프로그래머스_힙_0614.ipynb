## 더맵게

```python
import heapq
def solution(scoville, k):
    answer = 0
    heapq.heapify(scoville) # 기존 리스트 힙으로 변환
    while len(scoville) > 1:
        sco1 = heapq.heappop(scoville)
        sco2 = heapq.heappop(scoville)
        if sco1 < k or sco2 <k:
            heapq.heappush(scoville, sco1 + (sco2*2))
            answer += 1
        else:
            return answer
    if scoville[0]>=k:
        return answer
    else:
        return -1
    return answer 
```

## 이중우선순위큐

```python
# 풀이 1
import heapq

def solution(operations):
    answer = []
    for i in operations:
        a, b= i.split(' ')
        if a =="I":
            heapq.heappush(answer, int(b))
        else:
            if len(answer) ==0:
                pass
            elif b =="1":
                answer.pop(answer.index(heapq.nlargest(1, answer)[0]))
            else:
                heapq.heappop(answer) # 힙은 최솟값을 우선 순위로 삭제
        answer.sort()
    if not answer:
        return [0,0]
    else:
        return [heapq.nlargest(1, answer)[0],heapq.nsmallest(1, answer)[0] ]
                
                
```

```python
# 풀이2
def solution(operations):
    answer = []
    for i in operations:
        a, b = i.split(' ')
        if a== "I":
            answer.append(int(b))
        else:
            if len(answer) > 0:
                if b=='1':
                    answer.pop()
                else:
                    answer.pop(0)
            else:
                pass
        answer.sort()
    if not answer:
        return [0,0]
    else:
        return [max(answer), min(answer)]
```
