```python
def solution(A, B):
    for i in range(len(A)):
        for j in range(len(A[0])):
            A[i][j] += B[i][j]
    return A
solution([[1,2],[2,3]],[[3,4],[5,6]])
```

```python
def solution(A, B):
    answer = [[c+d for c, d in zip(a,b)] for a,b in zip(A, B) ]
    return answer
solution([[1,2],[2,3]],[[3,4],[5,6]])
```
