### 최대공약수 - 유클리드 호제법  
두 수의 최대 공약수 구하는 알고리즘,  
MOD(두 값을 나눈 나머지 구하는 연산)을 반복한다.  
1. 큰 수를 작은 수로 나눈 나머지를 구한다.  
2. 그 다음 나눴던 수와 나머지로 또 MOD연산한다.  
3. 반복한다. 나머지가 0이되기전까지  
예를 들면, 1071과 1029가 있다. 1071을 1029로 나눈 나머지는 42이다.
1029와 42로 나눈 나머지는 21이다. 42와 21로 나눈 나머지는 0이다. 최대공약수는 21이된다.
### 최소공배수  
x, y의 곱에서 x, y 최대공약수를 나누어 준 값과 같다.
### n개의 최소공배수  
2개의 최소공배수를 구하고 나머지 수 1개를 선택해 최소공배수를 구한다. 반복.

```python
# n개의 최소공배수
from math import gcd

def solution(arr):
    def lcm(x, y):
        return x * y // gcd(x, y)
    while True:
        arr.append(lcm(arr.pop(), arr.pop()))
        if len(arr)== 1:
            return arr[0]
```

```python
def gcd(x, y):
    while y: # y가 0이 될때까지 반복한다.
        # y를 x에 대입, x%y를 y에 대입
        if x < y:
            x, y = y, x
        x, y = y, x%y
    return x
def lcm(x, y):
    return x*y // gcd(x, y)

def solution(n,m):
    return gcd(n,m), lcm(n, m)
solution(3, 12)
```
