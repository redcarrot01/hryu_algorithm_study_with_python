```python
def solution(s):
    answer =''
    if len(s) %2 ==0:
        answer += s[len(s)//2-1: len(s)//2+1]
    else:
        answer += s[len(s)//2]
        
    return answer
solution("abcde")
```

```python
# 다른 사람 풀이
def solution(s):
    return s[(len(s)-1)//2 : len(s)//2+1]
solution("abcde")
```
