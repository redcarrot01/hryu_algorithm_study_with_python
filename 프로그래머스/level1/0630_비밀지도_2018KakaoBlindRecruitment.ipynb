```python
# 비트연산자 이용
def solution(n, arr1, arr2):
    answer = []
    for n1, n2 in zip(arr1, arr2):
        # 원소를 하나씩 꺼내와서 or 연산한 후, 2진법으로 바꾼다. 이때 0b를 지운다.
        b = str(bin(n1 | n2))[2:]
        b = '0'*(n- len(b))+b
        b = b.replace('0', ' ')
        b = b.replace('1', '#')
        answer.append(b)
    return answer
solution(5,[9, 20, 28, 18, 11] ,[30, 1, 21, 17, 28] )
```

```python
# 나머지 이용
def solution(n, arr1, arr2):
    answer = []
    
    for i in range(len(arr1)):
        result = ''
        b = arr1[i] | arr2[i] # 5|3 =7 이런 여산 가능
        for i in range(n):
            if b %2 ==0:
                result = ' '+ result
            else:
                result = '#'+ result
            b = b//2
        
        answer.append(result)
    return answer
solution(5,[9, 20, 28, 18, 11] ,[30, 1, 21, 17, 28] )
```
