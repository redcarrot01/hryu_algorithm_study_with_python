```python
def getDistance(value, n):
    if value == 0:
        value = 11
    if n == 0:
        n = 11
        
    x = (value-1)//3
    y = (value-1)%3
    n_x = n//3
    n_y = 1
    return abs(n_x-x)+ abs(n_y-y)
    
        
def solution(numbers, hand):
    left_location = [10]
    right_location = [12]
    result = []
    
    while numbers:
        n = numbers.pop(0)
        
        if n in (1, 4, 7):
            result.append('L')
            left_location.append(n)
        elif n in (3, 6, 9):
            result.append('R')
            right_location.append(n)
        else:
            left = left_location[-1]
            right = right_location[-1]
            leftDistance = getDistance(left, n)
            rightDistance = getDistance(right, n)
            
            if leftDistance < rightDistance:
                result.append('L')
                left_location.append(n)
            elif leftDistance > rightDistance:
                result.append('R')
                right_location.append(n)
            else:
                if hand == 'right':
                    result.append('R')
                    right_location.append(n)
                else:
                    result.append('L')
                    left_location.append(n)
    return ''.join(result)
            
solution([1, 3, 4, 5, 8, 2, 1, 4, 5, 9, 5],"right")    
solution([7, 0, 8, 2, 8, 3, 1, 5, 7, 6, 2], "left")
```
