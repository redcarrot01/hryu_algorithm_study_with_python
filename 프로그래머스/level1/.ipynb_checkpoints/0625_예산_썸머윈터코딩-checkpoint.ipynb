```python
def solution(d, budget):
    answer = 0
    cnt = 0
    d.sort()
    if d[0] > budget:
        return 0
    for i in range(len(d)):
        if budget == answer:
            return cnt
        if budget < answer:
            return cnt-1
        else:
            answer += d[i]
            cnt += 1
    return cnt
solution([1,3,2,5,4], 9)
```

```python
def solution(d, budget):
    d.sort()
    while budget < sum(d):
        d.pop()
    return len(d)
solution([1,3,2,5,4], 9)
```
