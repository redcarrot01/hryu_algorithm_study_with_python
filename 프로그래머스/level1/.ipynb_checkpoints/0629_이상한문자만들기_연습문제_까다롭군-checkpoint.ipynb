```python
def solution(s):
    answer = ''
    s_split = s.split(" ")
    for i in range(len(s_split)):
        ss =list(s_split[i])
        for j in range(len(ss)):
            if j%2==0: 
                ss[j]=ss[j].upper()
            elif j%2 != 0:
                ss[j]=ss[j].lower()
        s_split[i] = "".join(ss)
        answer += s_split[i] + " "
    return answer[:-1]

solution("try hello world")
```

```python
# 다른 풀이
def toWeirdCase(s):
    res = []
    for x in s.split(' '):
        word = ''
        for i in range(len(x)):
            c = x[i].upper() if i %2==0 else x[i].lower()
            word += c
        res.append(word)
    return ' '.join(res)
solution("try hello world")
```
