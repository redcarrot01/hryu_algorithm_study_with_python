```python
def correct_bracket(s):
    st, st1, st2, st3 = [], [], [], []
    if not s[0] in '(, {, [':
        return False
    for i in range(len(s)):
        if s[i] == '[':
            st1.append('[')
            st.append('[')
        elif s[i] == '{':
            st2.append('{')
            st.append('{')
        elif s[i] == '(':
            st3.append('(')
            st.append('(')
        elif s[i] == ']' and st1:
            st1.pop()
            if  st[-1] == '[':
                st.pop()
        elif s[i] == '}' and st2:
            st2.pop()
            if  st[-1] == '{':
                st.pop()
        elif s[i] == ')' and st3:
            st3.pop()
            if  st[-1] == '(':
                st.pop()
  
    if len(st1)==0 and len(st2)==0 and len(st3)==0 and len(st)==0:
        return True
    else: 
        return False

def solution(s):
    answer = 0
    turn_s = []
    for i in range(len(s)):
        if i ==0:
            if correct_bracket(s):
                answer += 1
        else:
            turn_s = s[i:] + s[:i]
            
            if correct_bracket(turn_s):
                answer += 1
    return answer

print(solution("[](){}"))
print(solution("}]()[{"))
print(solution("[)(]"))
print(solution("}}}"))
print(solution("[(])"))

# 테스트 1 〉	통과 (116.47ms, 10.3MB)
# 테스트 2 〉	통과 (130.88ms, 10.4MB)
# 테스트 3 〉	통과 (131.50ms, 10.1MB)
# 테스트 4 〉	통과 (129.96ms, 10.3MB)
# 테스트 5 〉	통과 (136.39ms, 10.3MB)
# 테스트 6 〉	통과 (139.40ms, 10.4MB)
# 테스트 7 〉	통과 (138.62ms, 10.3MB)
# 테스트 8 〉	통과 (136.75ms, 10.3MB)
# 테스트 9 〉	통과 (137.58ms, 10.2MB)
# 테스트 10 〉	통과 (138.59ms, 10.2MB)
# 테스트 11 〉	통과 (137.06ms, 10.2MB)
# 테스트 12 〉	통과 (0.01ms, 10.3MB)
# 테스트 13 〉	통과 (0.02ms, 10.2MB)
# 테스트 14 〉	통과 (0.02ms, 10.2MB)
# 채점 결과
# 정확성: 100.0
# 합계: 100.0 / 100.0
```
