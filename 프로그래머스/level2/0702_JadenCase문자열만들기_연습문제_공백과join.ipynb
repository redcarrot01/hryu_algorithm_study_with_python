split : 공백도 리스트의 원소로 취급
~~~python
s = 'erer i   e'
ss = s.split(' ')
' '.join(ss)
# 'erer i   e'
~~~

```python
def solution(s):
    s_split = s.split(' ')
    print(s_split)
    for i in range(len(s_split)):
        s_split[i] = s_split[i].capitalize()
    return ' '.join(s_split)
solution("2v 3hello    m y friend 23HIz")
```
