```python
def solution(rows, columns, queries):
    answer = []
    graph = [[0 for i in range(columns)] for j in range(rows)]
    
    # 초기 그래프
    idx = 1
    for i in range(rows):
        for j in range(columns):
            graph[i][j] = idx
            idx += 1
            
    for x1, y1, x2, y2 in queries:
        first = graph[x1-1][y1-1]
        mini = first
        # 왼쪽 세로 부분 (아래에서 위로 진행)
        for k in range(x1-1, x2-1):
            temp = graph[k+1][y1-1]
            graph[k][y1-1] = temp
            mini = min(mini, temp)
        # 아래 가로 부분
        for k in range(y1-1,y2-1):
            temp = graph[x2-1][k+1]
            graph[x2-1][k] =  temp
            mini = min(mini,  temp)
        # 오른쪽 세로 부분 
        for k in range(x2-1,x1-1,-1):
            temp = graph[k-1][y2-1]
            graph[k][y2-1] =  temp
            mini = min(mini,  temp)
        # 위 가로 부분
        for k in range(y2-1,y1-1,-1):
            temp = graph[x1-1][k-1]
            graph[x1-1][k] =  temp
            mini = min(mini,  temp)
        
        graph[x1-1][y1] = first
        answer.append(mini)
    #print(graph)
    return answer
solution(6,6,[[2,2,5,4],[3,3,6,6],[5,1,6,3]])
```
