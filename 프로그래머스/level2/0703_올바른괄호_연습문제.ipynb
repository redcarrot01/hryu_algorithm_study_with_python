```python
def solution(s):
    answer = True
    stack = []
    for i in range(len(s)):
        if s[i] == '(':
            stack.append(s[i])
        elif len(stack)!= 0 and s[i] == ')':
            stack.pop()
        elif len(stack) == 0 and s[i] == ')':
            return False
    if len(stack) == 0:
        return True
    else:
        return False
solution("(())()")
```
