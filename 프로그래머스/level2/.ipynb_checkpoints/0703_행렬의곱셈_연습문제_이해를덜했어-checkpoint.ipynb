```python
행렬의 곱셈에서 arr1의 각 행의 길이 = arr2의 각 열의 길이  
arr1 각 행은 arr2의 모든 열을 돌아야한다.  
열을 돌때, arr2의 1개 열을 처리하기 위해서, 특정 열의 모든 행을 돌아야 한다.
어렵다 진짜
```

```python
def solution(arr1, arr2):
    answer = []
    row = len(arr1)
    col = len(arr2[0])
    for i in range(row):
        result_list = []
        for j in range(col):
            result = 0
            for k in range(len(arr2)):
                result += arr1[i][k]*arr2[k][j]
            result_list.append(result)
        answer.append(result_list)
    return answer
```
