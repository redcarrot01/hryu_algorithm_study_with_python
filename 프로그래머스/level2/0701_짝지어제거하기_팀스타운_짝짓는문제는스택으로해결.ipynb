```python
# 삭제하고, 정렬하는 시간이 있어서 오래 걸린다.
# while , for문을 사용, 앞뒤 비교하는 것은 시간초과가 발생 -> 문자열 길이가 1,000,000이하
def solution(s):
    s = list(s)
    while s:
        for i in range(1, len(s)):
            if s[i-1] == s[i]:
                s = s[:i-1]+s[i+1:]
                break
        else: # for문에서 break가 동작하지 않은 경우는 else로 빠져나옴
            break
                
    if len(s)==0:
        return 1
    else:
        return 0
solution('baabaa')
#solution('cdcd')
```

```python
# 스택 활용하는 문제다.
def solution(s):
    answer = []
    for i in s:
        if answer:
            if answer[-1] == i:
                answer.pop()
            else:
                answer.append(i)
        else:
            answer.append(i)
    if answer:
        return 0
    else:
        return 1
```
