```python
def solution(relation):
    answer = 0
    col = len(relation[0])
    row = len(relation)
    cnt = 0
    for j in range(col):
        temp = []
        for i in range(row):
            temp.append(relation[i][j])
            
        if len(set(temp)) == row:
            cnt+= 1
            print(cnt)
            for k in range(row):
                del(relation[k][j])
                break
            
        
            
        
    return answer

solution([["100","ryan","music","2"],["200","apeach","math","2"],["300","tube","computer","3"],["400","con","computer","4"],["500","muzi","music","3"],["600","apeach","music","2"]])
```
