```python
1. board를 배열로 변환
2. 전체 보드를 탐색
- 똑같은 블록 4개 위치한 곳 찾기
- 찾았다면 4개 블록의 인덱스를 다른 리스트에 저장
- 현재 동일한 블록들이 터진 것으로 바꿈
- 터진 블록 중 위에 있는 블록을 밑으로 내려줌
- 다시 전체 보드 탐색
```

```python
def solution(m, n, board):
    answer = 0
    graph = [[0] for i in range(len(board))]
    for i in range(len(board)):
        graph[i] = list(board[i])
    #print(graph)
    while True:
        check = []
        for i in range(n-1):
            for j in range(m-1):
                if graph[i][j] == graph[i+1][j] == graph[i][j+1] == graph[i+1][j+1] != 0:
                    check.append((i+1, j+1))
                    check.append((i+1, j))
                    check.append((i, j+1))
                    check.append((i, j))
        if len(check) != 0:
            answer += set(check)
            for i in check:
                graph[i[0]][i[1]] = '0'
                
            for i in check:
                check_n = i[0] -1
                put_n = i[0]
                
                while check_n >= 0:
                    # 만약 위에는 블록이 있고
                    if graph[put_n][i[1]] == "0" and graph[check_n][i[1]] != "0":
                        graph[put_n][i[1]] = graph[check_n][i[1]] # 내려주기
                        graph[check_n][i[1]] = '0'
                        put_n -= 1
                    check_n -= 1
        else:
            break
    return answer

solution(4, 5, ["CCBDE", "AAADE", "AAABF", "CCBBF"])
```
