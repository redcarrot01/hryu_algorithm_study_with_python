### 나의 풀이  
1. 문자를 두글자씩 끊어서 다중 집합을 만들어야 한다. 이때, 입력이 문자열로 주어지기 때문에 리스트를 생성하여 2글자씩 넣어준다.`(중복된 원소를 포함해야 하므로 set을 쓰지 않고 리스트를 사용한다.)`  

> 단, 이때 문자열이 알파벳인지 확인하고, 그렇다면 대소문자 비교를 하지 않기 위해 소문자로 변환하여 리스트에 넣어준다.  

2. 두 리스트의 교집합과 합집합을 구하는 과정이다.  

문자열의 길이가 짧은 쪽을 기준으로 반복문을 돌린다.   

> 예시 : A = {1, 1, 2, 2, 3}, 다중집합 B = {1, 2, 2, 4, 5}라고 하면, 교집합 A ∩ B = {1, 2, 2}, 합집합 A ∪ B = {1, 1, 2, 2, 3, 4, 5}  

~~~python
 if len(list1) > len(list2):

~~~

- 교집합 구하는 과정
   
  `A∩B, B =B-A`
  A를 기준으로 반복문을 돌리면, A와 B에 같은 원소가 있는 경우, B에 있는 원소를 삭제하고, 공통 원소는 교집합에 추가한다.    
  (사실 교집합에는 원소 말고 None 타입이 추가된다. 교집합의 개수만 구하는 것이므로 None 타입의 개수를 알면 문제 해결은 가능하다)     
  A = {1, 1, 2, 2, 3} B={4, 5} 교집합 = {1, 2, 2}        

- 합집합 구하는 과정  
`AUB = (B-A) + A`
B 리스트는 A와 서로소 집합 관계로, B-A이며, 이것을 합집합에 추가한다.    
A를 합집합에 추가한다.    
합집합 = B = {4, 5} , 합집합 += A = {1, 1, 2, 2, 3, 4, 5}    
  

3. 두 리스트의 길이를 구한 후, 조건에 맞게 계산해주면 된다.    

```python
from collections import defaultdict
def solution(str1, str2):
    answer = 0

    list1 = []
    list2 = []
    union = []
    intersection=[]
    
    for i in range(len(str1)-1):
        if str1[i].isalpha() and str1[i+1].isalpha():
            list1.append(str1[i:i+2].lower())

    for i in range(len(str2)-1):
        if str2[i].isalpha() and str2[i+1].isalpha():
            list2.append(str2[i:i+2].lower())
            

    if len(list1) > len(list2):
        intersection = [list1.remove(i) for i in list2 if i in list1]
        union = list1
        union += list2
    else:
        intersection = [list2.remove(i) for i in list1 if i in list2]
        union = list2
        union += list1
    
    #print(intersection, union)
    if len(intersection) ==0 and len(union) ==0:
        return 1*65536
    return int((len(intersection)/len(union))*65536)
# solution("FRANCE",'french')
# solution("aa1+aa2", "AAAA12")
# #solution("handshake","shake hands")
#solution("aaabbbb", "aaaabbb")
solution("aabbc", "abbde")
```

### 교집합, 합집합 부분 다른 풀이

```python
def solution(str1, str2):
    
    list1 = [str1[i:i+2].lower() for i in range(len(str1)-1) if str1[i:i+2].isalpha()]
    list2 = [str2[i:i+2].lower() for i in range(len(str2)-1) if str2[i:i+2].isalpha()]
    
    list_ = set(list1)|set(list2)
    intersection = []
    union = []
    
    if list_:
        for i in list_ :
            intersection.extend([i]*min(list1.count(i),list2.count(i)))
            union.extend([i] * max(list1.count(i), list2.count(i)))
        
        answer = int(len(intersection) / len(union) * 65536)
        return answer
    else:
        return 65536
solution("aaabbbb", "aaaabbb")
```

### Collections 모듈의 Counter 클래스 사용법  

데이터 개수를 셀 때 사용하는 파이썬의 collections 모듈의 Counrer 클래스  
Counter 클래스는 사전을 확장하고 있는 클래스로 사전 API 다 쓸 수 있다.  
~~~python
from collections import Counter
Counter('hello world') # Counter({'h': 1, 'e': 1, 'l': 3, 'o': 2, ' ': 1, 'w': 1, 'r': 1, 'd': 1})
~~~
주어진 단어에서 가장 많이 사용하는 알파벳과 그 개수 구하기  
~~~python
from collections import Counter
Counter('hello world').most_common(1) # [('l', 3)
Counter('hello world').most_common() # 개수 기준으로 내림차순된 리스트가 출력된다. 
~~~
값 = 개수 상태로 입력이 가능하다.
~~~python
import collections
c = collections.Counter(a=2, b=3, c=2)
print(collections.Counter(c))
print(sorted(c.elements()))
# Counter({'b': 3, 'a': 2, 'c': 2})
# ['a', 'a', 'b', 'b', 'b', 'c', 'c']
~~~

### Counter 이용한 풀이

```python
def solution(str1, str2):
    str1 = str1.lower()
    str2 = str2.lower()
    
    list1 = [str1[i:i+2] for i in range(len(str1)-1) if str1[i:i+2].isalpha()]
    list2 = [str2[i:i+2] for i in range(len(str2)-1) if str2[i:i+2].isalpha()]
    
    Counter1 = Counter(list1)
    Counter2 = Counter(list2)
    intersection = list((Counter1 & Counter2).elements())
    union = list((Counter1 | Counter2).elements())
    
    if len(union) ==0:
        return 65536
    else:
        return int(len(intersection) / len(union) * 65536)
solution("FRANCE",'french')
```

### 정규표현식 이용한 풀이

- [a-zA-Z] : 알파벳이 있는 문자 ['aa', 'a1', '+a', 'aa', 'a2']  
- [^a-zA-Z] : 알파벳이 아닌 문자가 있는 문자 ['a1', '1+', '+a', 'a2']  
- if not re.findall('[^a-zA-Z]+') : 알파벳이 아닌 문자가 아닌 문자가 1개 이상. all 알파벳 
- if not re.findall('[^a-zA-Z]') : 알파벳이 아닌 문자가 아닌 문자 1개   

```python
import re
import math

def solution(str1, str2):
    str1 = [str1[i:i+2].lower() for i in range(len(str1)-1) if not re.findall('[^a-zA-Z]', str1[i:i+2])] 
    str2 = [str2[i:i+2].lower() for i in range(len(str2)-1) if not re.findall('[^a-zA-Z]', str2[i:i+2])]
    intersection = set(str1) & set(str2)
    union = set(str1) | set(str2)
    print(str1, str2)
    if len(union) ==0:
        return 65536
    
    inter_sum = sum([min(str1.count(i), str2.count(i)) for i in intersection ])
    union_sum = sum([max(str1.count(i), str2.count(i)) for i in union])
    #print(inter_sum, union_sum)
    
    if union_sum==0:
        return 65536
    else:
        return int(inter_sum/union_sum * 65536)
solution("aa1+aa2", "AAAA12")
```
