```python
# LRU : 가장 오랫동안 참조되지 않은 페이지 교체
# cache hit : 캐시에 있는 데이터 불러오는 경우 
# cache miss : 캐시에 없는 데이터 불러오는 경우


def solution(cacheSize, cities):
    answer = 0
    cache = []*cacheSize
    
    for city in cities:
        city = city.lower()
        if city in cache:
            cache.remove(city)
            cache.append(city)
            answer += 1
        else:
            cache.append(city)
            answer += 5
            if len(cache) > cacheSize:
                cache.pop(0)
        
    return answer
solution(3, ["Jeju", "Pangyo", "Seoul", "NewYork", "LA", "Jeju", "Pangyo", "Seoul", "NewYork", "LA"])

테스트 1 〉	통과 (0.01ms, 10.2MB)
테스트 2 〉	통과 (0.01ms, 10.1MB)
테스트 3 〉	통과 (0.01ms, 10.1MB)
테스트 4 〉	통과 (0.01ms, 10.2MB)
테스트 5 〉	통과 (0.01ms, 10.1MB)
테스트 6 〉	통과 (0.01ms, 10.2MB)
테스트 7 〉	통과 (0.01ms, 10.2MB)
테스트 8 〉	통과 (0.01ms, 10.2MB)
테스트 9 〉	통과 (0.01ms, 10.2MB)
테스트 10 〉	통과 (0.05ms, 10.2MB)
테스트 11 〉	통과 (62.34ms, 17.6MB)
테스트 12 〉	통과 (0.05ms, 10.2MB)
테스트 13 〉	통과 (0.09ms, 10.1MB)
테스트 14 〉	통과 (0.14ms, 10.2MB)
테스트 15 〉	통과 (0.19ms, 10.2MB)
테스트 16 〉	통과 (0.31ms, 10.2MB)
테스트 17 〉	통과 (0.47ms, 10.2MB)
테스트 18 〉	통과 (0.81ms, 10.1MB)
테스트 19 〉	통과 (0.77ms, 10.3MB)
테스트 20 〉	통과 (1.02ms, 10.2MB)
```

```python
# 다른 풀이 - deque 사용
def solution(cacheSize, cities):
    import collections
    cache = collections.deque(maxlen=cacheSize)
    time = 0
    for i in cities:
        s = i.lower()
        if s in cache:
            cache.remove(s)
            cache.append(s)
            time += 1
        else:
            cache.append(s)
            time += 5
    return time
solution(3, ["Jeju", "Pangyo", "Seoul", "NewYork", "LA", "Jeju", "Pangyo", "Seoul", "NewYork", "LA"])

테스트 1 〉	통과 (0.01ms, 10.3MB)
테스트 2 〉	통과 (0.01ms, 10.2MB)
테스트 3 〉	통과 (0.01ms, 10.3MB)
테스트 4 〉	통과 (0.01ms, 10.1MB)
테스트 5 〉	통과 (0.01ms, 10.2MB)
테스트 6 〉	통과 (0.01ms, 10.2MB)
테스트 7 〉	통과 (0.01ms, 10.2MB)
테스트 8 〉	통과 (0.02ms, 10.3MB)
테스트 9 〉	통과 (0.01ms, 10.2MB)
테스트 10 〉	통과 (0.05ms, 10.2MB)
테스트 11 〉	통과 (60.76ms, 17.4MB)
테스트 12 〉	통과 (0.03ms, 10.3MB)
테스트 13 〉	통과 (0.06ms, 10.2MB)
테스트 14 〉	통과 (0.09ms, 10MB)
테스트 15 〉	통과 (0.16ms, 10.2MB)
테스트 16 〉	통과 (0.18ms, 10.2MB)
테스트 17 〉	통과 (0.24ms, 10.3MB)
테스트 18 〉	통과 (0.53ms, 10.2MB)
테스트 19 〉	통과 (0.79ms, 10.2MB)
테스트 20 〉	통과 (1.05ms, 10.2MB)
```
