```python
from collections import defaultdict
import copy
def solution(str1, str2):
    answer = 0

    dic1 = []
    dic2 = []
    union = []
    intersection=[]
    
    for i in range(len(str1)-1):
        if str1[i].isalpha() and str1[i+1].isalpha():
            dic1.append(str1[i:i+2].lower())

    for i in range(len(str2)-1):
        if str2[i].isalpha() and str2[i+1].isalpha():
            dic2.append(str2[i:i+2].lower())
            

    if len(dic1) > len(dic2):
        intersection = [dic1.remove(i) for i in dic2 if i in dic1]
        union = dic1
        union += [x for x in dic2 if x not in intersection]
    else:
        intersection = [dic2.remove(i) for i in dic1 if i in dic2]
        union = dic2
        union += [x for x in dic1 if x not in intersection]
    

    if len(intersection) ==0 and len(union) ==0:
        return 1*65536
    return int((len(intersection)/len(union))*65536)
solution("FRANCE",'french')
solution("aa1+aa2", "AAAA12")
#solution("handshake","shake hands")
solution("aaabbbb", "aaaabbb")
```
