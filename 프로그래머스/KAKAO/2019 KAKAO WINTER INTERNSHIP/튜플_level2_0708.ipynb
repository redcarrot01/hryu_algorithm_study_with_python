```python
split() : 구분자 기준으로 문자열 쪼개서 리스트로 반환 
```

```python
def solution(s):
    s = s[2:-2].split("},{") # 문자열 중간에 '},{' 으로 구분되어 있고, 이것을 기준으로 쪼갠다.
    result = []
    s.sort(key= lambda x : len(x) )
    for i in range(len(s)):
        ss = s[i].split(",")
        result.append(set(ss))

    answer = [list(result[0])[0]]
    for j in range(1, len(result)):
        answer.append(list(result[j] - result[j-1])[0])

    final = []
    for i in answer:
        final.append(int(i))
    return final
```

```python
# 다른 풀이
def solution(s):
    answer = []

    s1 = s.lstrip('{').rstrip('}').split('},{')

    new_s = []
    for i in s1:
        new_s.append(i.split(','))

    new_s.sort(key = len)

    for i in new_s:
        for j in range(len(i)):
            if int(i[j]) not in answer:
                answer.append(int(i[j]))

    return answer
```

```python
# ㅇㅇ
def solution(s):
    # {{, }}를 제거 후 },{ 으로 나누기
    data = s[2:-2].split("},{")
    # 길이 별로 오름차순 정렬
    data = sorted(data, key=lambda x: len(x))
    answer = []
    for item in data:
        # 각각의 원소로 분류 후
        item = list(map(int, item.split(",")))
        for value in item:
            # 포함되어 있지 않으면 input
            if value not in answer:
                answer.append(value)
    return answer
```
