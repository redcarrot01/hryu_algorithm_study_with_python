배열을 r만큼 반시계 방향으로 돌리기..  
min(N, M) mod 2 = 0 -> 가장 작은 값이 항상 2의 배수란 의미.. 3x4 라면 회전불가한 부분이 생긴다.  
1. 테두리 개수 세기 -> cnt = min(n,m) // 2  
2. 테두리 개수 만큼 회전해야하므로 반복문    
- 회전 시작 하는 지점은 다른 변수에 저장   
- 다른 곳은 반시계 방향 회전    
- 회전 시작 지점 회전 반영  

```python
n, m, r = map(int, input().split())
graph = [list(map(int, input().split())) for _ in range(n)]

def solution():
    
    for _ in range(r):
        rotate()
    for i in graph:
        for j in i:
            print(j, end=" ")
        print()
        
 # 5 * 4            
def rotate():
    cnt = min(n,m) // 2 # 2 
    for i in range(cnt): # (0, 1)
        nn = n- i -1 # 5 -1 = 4
        mm = m -i -1 # 4- 1 =3
        start = graph[i][i] # 
        # 위쪽 가로 : 왼 <- 오
        for j in range(i , mm):
            graph[i][j] = graph[i][j+1]
        # 오른쪽 세로 : 아래 <- 위
        for j in range(i, nn):
            graph[j][mm] = graph[j+1][mm]
        # 아래쪽 가로 : 왼 -> 오
        for j in range(mm, i, -1):
            graph[nn][j] = graph[nn][j-1]
        # 왼쪽 세로 : 위 -> 아래
        for j in range(nn, i, -1):
            graph[j][i] = graph[j-1][i]
        graph[i+1][i] = start 
solution()   
```
