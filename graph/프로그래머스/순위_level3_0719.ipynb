모든 노드가 모든 노드에 대해 승패 여부를 탐색해야 한다.   
플로이드 워셜 알고리즘을 이용했다.  
최단 경로 대신 승패 여부를 기록한다.  
모든 노드에서 모든 노드로 승패 여부를 담기 때문에 이차 리스트를 이용한다.  
승패 여부는 a > b > c : a > c 이며, graph[a][c] 1을 기록, c < a 이므로  graph[c][a] -1을 기록      
graph[i][i] 는 자기자신이므로 0이며, 각 노드에 대해 0이 하나라면 승패가 모두 결정된 것이므로 카운트   

```python
def solution(n, results):
    answer = 0
    graph = [[0]*(n) for _ in range(n)]
    for a, b in results:
        graph[a-1][b-1] = 1
        graph[b-1][a-1] = -1
    
    for k in range(n):
        for i in range(n):
            for j in range(n):
                if graph[i][j] == 0:
                    if graph[i][k] == 1 and graph[k][j] == 1:
                        graph[i][j] =1
                    elif graph[i][k] == -1 and graph[k][j] == -1:
                        graph[i][j] = -1
    for i in range(n):
        if graph[i].count(0) == 1:
            answer += 1
    return answer
solution(5,[[4, 3], [4, 2], [3, 2], [1, 2], [2, 5]])
```
