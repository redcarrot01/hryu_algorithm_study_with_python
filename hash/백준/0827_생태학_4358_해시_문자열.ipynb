```python
import sys
d = dict()
cnt =0
for i in sys.stdin:
    if i == '\n':
        break
    t = i.rstrip()
    cnt += 1
    if t in d:
        d[t] += 1
    else:
        d[t] = 1
sorted_d = sorted(d.items(), key = lambda x : x[0])
for i, j in sorted_d:
    divide = (j/cnt) * 100
    rate = round(divide,4)
    #print(i, float(rate))
    print(i, '%.4f' %rate)
```

```python
import sys
trees = dict()
info = sys.stdin.read()
data = info.split('\n')

for i in data[:-1]:
    if i in trees:
        trees[i] += 1
    else:
        trees[i] =1
    cnt += 1

for key in sorted(list(trees.keys())):
    print(key, "%.4f"%(trees[key]/cnt*100))
```
