```python
n = int(input())
s = list(map(str, input()))
dp = [1e9]*(n+1)

dp[0] = 0
for i in range(0, n-1): # 
    if s[i] == 'B':
        for j in range(i+1, n):
            if s[j] == 'O':
                dp[j] = min(dp[j],(j-i)*(j-i) + dp[i])

    elif s[i] == 'O':
        for j in range(i+1,n ):
            if s[j] == 'J':
                dp[j] = min(dp[j],(j-i)*(j-i) + dp[i])

    else :
        for j in range(i+1, n):
            if s[j] == 'B':
                dp[j] = min(dp[j],(j-i)*(j-i) + dp[i])

if dp[n-1] == 1e9:
    print(-1)
else:
    print(dp[n-1])

```
