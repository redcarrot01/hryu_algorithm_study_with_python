```python
n = int(input())
arr = list(map(int, input().split())) 
arr2 = arr[::-1]
print(arr)
dp_r = [1]*n
dp_l = [1]*n
answer = [0]*n
for i in range(n):
    for j in range(i):
        if arr[i] > arr[j]:
            dp_r[i] = max(dp_r[i], dp_r[j]+1)
        if arr2[i] > arr2[j]:
            dp_l[i] = max(dp_l[i], dp_l[j]+1)
for i in range(n):
    answer[i] = dp_r[i] + dp_l[n-1-i]-1

print(max(answer))
```
