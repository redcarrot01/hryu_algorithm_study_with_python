`n은 n-1인 수에서 1의자리에 +1, -1이 온다는 것을 알 수 있다.`

예를 들어, n=3에서 121이라는 수가 있다면, n=4인 수를 만드는 방법이 

1210, 1212 이런식으로 십의자리숫자에 +1, -1한 수를 1의자리에 붙여준다는 것을 알 수 있다.

> dp [n] [m] = dp[n-1] [m-1] + dp[n-1] [m+1]
>
> 수의 길이가 n이고, 1의 자리에 m인 경우의 계단 수

m =0인 경우 1의자리에 +1인 1만 올 수 있고,

m =9인 경우 1의 자리에 -1인 8만 올 수 있다.

m이 1~8인 경우 +1, -1 다 올 수 있다.

```python
n = int(input())
dp = [[0 for i in range(10)] for j in range(101)]
for i in range(1, 10):
    dp[1][i] = 1
for i in range(2, n+1):
    for j in range(10):
        if j == 0:
            dp[i][j] = dp[i-1][1]
        elif j == 9:
            dp[i][j] = dp[i-1][8]
        else:
            dp[i][j] = dp[i-1][j-1] + dp[i-1][j+1]
print(sum(dp[n]) % 1000000000)
```
