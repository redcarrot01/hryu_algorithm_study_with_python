```python
n = int(input())
arr = list(map(int, input().split()))
dp = [0]*n
dp[0] = arr[0]
for i in range(n):
    for j in range(i):
        if arr[i] > arr[j]:
            if dp[j] > arr[j]:
                dp[i] = max(dp[j] + arr[i], dp[i])
            else:
                dp[i] = max(arr[j] + arr[i], dp[i])
print(max(dp))
            
```
