```python
n, m = map(int, input().split())

arr = [0]*n
for i in range(n):
    arr[i] = int(input())

name = [0]*1001
dp =[ [-1]* 1001 for _ in range(1001)] # i열에서 j번째 사람 이름을 썼을 떄 총 칸수 제곱합

def solve(pos, num):
    if num == n: 
        return 0
    
    result = dp[pos][num]
    if result != -1:
        return result
    
    nn = m-pos +2
    result = (solve(name[num]+2, num+1) + nn*nn); # 다음행으로 넘어감
    
    if pos + name[num] -1 <= m:
        result = min(result, solve(pos + name[num]+1, num+1))

    return result
print(solve(name[0]+1, 1))
```

```python
import sys
sys.setrecursionlimit(9999999)
n,m=map(int,input().split())
name=[]
dp=[[-1]*1000 for _ in range(1000)] #1000x1000리스트 생성
for _ in range(n):
    name.append(int(input())) #이름 길이 입력받음
    

def func(num,col): 
# num은 이번에 쓸 이름의 번호, col은 쓸 열의 번호를 입력받아
# 그 상황에서의 남은 공백제곱의 최소값을 출력하는 함수
    if num==n: return 0 #마지막 이름까지 다쓰면 0 리턴
    elif dp[num][col]!=-1: return dp[num][col] #메모이제이션
    else:
        if name[num]+col<m: left=func(num+1,col+name[num]+1) 
            #num+1(다음 번째 이름), col+name[num]+1은 이름을 추가적으로 쓰고 공백1칸
            #col(이번부터 쓸 열)+이번에 쓸 이름 길이가 가로폭이 넘지 않으면 계속 씀
        elif name[num]+col==m: left=func(num+1,col+name[num])
            #col(이번부터 쓸 열)+이번에 쓸 이름 길이가 같으면 마지막에는 이름에 공백을 붙이지 않음
        else: left=func(num+1,name[num]+1)+(m-col+1)**2
            #그렇지않고 가로폭 초과시 강제로 엔터(개행)하면서 남은 공백값 제곱도 더함
       
        right=func(num+1,name[num]+1) + (m-col+1)**2
            #남은 칸상관없이 강제 개행하면서 공백제곱 더함
        
        dp[num][col]=min(left,right) #둘의 최소값 저장
        return dp[num][col]#리턴

print(func(1,name[0]+1)) #첫째줄에 이름 하나 쓰고 시작
```
