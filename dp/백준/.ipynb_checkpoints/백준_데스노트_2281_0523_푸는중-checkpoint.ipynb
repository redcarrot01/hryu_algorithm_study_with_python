남의코드도 이해가 잘 안가성 힘드러썯ㅇ
n, m : 1~1000

```python
n, m = map(int, input().split())
p = [0]*n
for i in range(n):
    p[i] = int(input())

d = [1e9] * 1001

def go(i):
    result = d[i]
    if result != 1e9:
        return result
    
    temp = list(range(i, n, 1))
    sum_num = list(range(p[i], m+1, p[temp[i]]+1))
                   
    for j,sum_j in zip(temp, sum_num):
        if j == n-1:
           result = 0
        else:
           rest = m - sum_j
           result = min(result, rest*rest + go(j+1))
    return result
               

go(0)
```
