n : 3 ~100
dp[i],  i : 0~9, dp값은 0~20
dp[i][j] : i번 계산했을 때 숫자 j값이 나올 수 있는 경우의 수
점화식
j - arr[i] >= 0인경우 dp[i][j-arr[i]]+= dp[i-1][j]
j + arr[i] <= 20인경우 dp[i][j+arr[i]] += dp[i-1][j]

1. 0번 인덱스의 arr[0]값 카운트
dp[0][arr[0]] += 1
2. 1~ n-2까지 연산, 조건에 만족하면 이전 경우의 수dp값을 누적
j - arr[i] >= 0인경우 dp[i][j-arr[i]]+= dp[i-1][j]
j + arr[i] <= 20인경우 dp[i][j+arr[i]] += dp[i-1][j]
3. 결과
dp[n-2][arr[n-1]] 
등식의개수 n-1이며, 마지막 등호(=) 제외한다

```python
n = int(input())
arr = list(map(int, input().split()))
dp = [[0]*21 for _ in range(n+1)]

dp[0][arr[0]] += 1
for i in range(1, n-1):
    for j in range(21):
        if dp[i-1][j]:
            if j - arr[i] >= 0: dp[i][j-arr[i]]+= dp[i-1][j]
            if j + arr[i] <= 20: dp[i][j+arr[i]] += dp[i-1][j]
print(dp[n-2][arr[n-1]])
```
