```python
t = int(input())
for _ in range(t):
    n = int(input())
    dp = [list(map(int, input().split())) for _ in range(2)]
    
    dp[0][1] += dp[1][0]
    dp[1][1] += dp[0][0]
    for i in range(2, n):
        for j in range(2):
            if j ==0 :
                if dp[1][i-1] > dp[1][i-2]:
                    dp[0][i] += dp[1][i-1]
                else: 
                    dp[0][i] += dp[1][i-2]
            elif j==1:
                if dp[0][i-1] > dp[0][i-2]:
                    dp[1][i] += dp[0][i-1]
                else: 
                    dp[1][i] += dp[0][i-2]
                    
    print(max(dp[0][n-1], dp[1][n-1]))
```
