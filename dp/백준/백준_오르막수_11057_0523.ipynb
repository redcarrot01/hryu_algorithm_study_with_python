```python
# 1. 중복 조합
import math
n = int(input())
result = math.comb(10+n-1, n)
print(result % 10007)
```

```python
# 2. 정석 풀이
n = int(input())

dp = [[0]*10 for _ in range(n+1)]

# n =1일때
dp[1] = [1] * 10

for i in range(2, n+1):
    dp[i][0] = 1
    
    for j in range(10):
        dp[i][j] = dp[i-1][j] + dp[i][j-1]
print(sum(dp[n])% 10007)   
```
